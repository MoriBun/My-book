## 3.1 Chuẩn (Norms)

### Lý thuyết

**Định nghĩa 3.1:** Một **chuẩn** trên không gian vectơ $V$ là hàm $\|\cdot\|: V \to \mathbb{R}$ gán cho mỗi vectơ $\mathbf{x}$ một độ dài $\|\mathbf{x}\| \in \mathbb{R}$, thỏa mãn với mọi $\lambda \in \mathbb{R}$ và $\mathbf{x}, \mathbf{y} \in V$:

1. **Thuần nhất tuyệt đối (Absolutely homogeneous):** $\|\lambda \mathbf{x}\| = |\lambda| \|\mathbf{x}\|$
2. **Bất đẳng thức tam giác (Triangle inequality):** $\|\mathbf{x} + \mathbf{y}\| \leq \|\mathbf{x}\| + \|\mathbf{y}\|$
3. **Xác định dương (Positive definite):** $\|\mathbf{x}\| \geq 0$, và $\|\mathbf{x}\| = 0 \iff \mathbf{x} = \mathbf{0}$

> **Bất đẳng thức tam giác** có ý nghĩa hình học trực tiếp: trong bất kỳ tam giác nào, tổng độ dài hai cạnh luôn lớn hơn hoặc bằng độ dài cạnh còn lại.

### Các chuẩn quan trọng

**Chuẩn Manhattan ($\ell_1$):**

$$\|\mathbf{x}\|_1 := \sum_{i=1}^n |x_i| \tag{3.3}$$

Tổng giá trị tuyệt đối của các thành phần. Tập $\{\mathbf{x} \in \mathbb{R}^2 : \|\mathbf{x}\|_1 = 1\}$ có hình thoi (hình vuông xoay 45°).

**Chuẩn Euclid ($\ell_2$):**

$$\|\mathbf{x}\|_2 := \sqrt{\sum_{i=1}^n x_i^2} = \sqrt{\mathbf{x}^\top \mathbf{x}} \tag{3.4}$$

Khoảng cách Euclid từ $\mathbf{x}$ đến gốc tọa độ. Tập $\{\mathbf{x} \in \mathbb{R}^2 : \|\mathbf{x}\|_2 = 1\}$ có hình tròn đơn vị.

> **Quy ước:** Trong suốt cuốn sách này, nếu không nói gì thêm, chuẩn mặc định được dùng là **chuẩn Euclid** $\|\cdot\|_2$.

### Ví dụ trực quan

Cho $\mathbf{x} = [3, 4]^\top \in \mathbb{R}^2$:
- $\|\mathbf{x}\|_1 = |3| + |4| = 7$ (khoảng cách đi theo đường kẻ ô)
- $\|\mathbf{x}\|_2 = \sqrt{9 + 16} = \sqrt{25} = 5$ (khoảng cách đường chim bay)

Hai chuẩn đo "độ dài" theo cách khác nhau — $\ell_1$ phù hợp với di chuyển theo lưới (ví dụ đường phố Manhattan), $\ell_2$ phù hợp với khoảng cách hình học thông thường.

## 3.2 Tích Trong (Inner Products)

### Lý thuyết

Một **ánh xạ song tuyến tính** (bilinear mapping) $\Omega: V \times V \to \mathbb{R}$ nhận hai vectơ và trả về một số thực, đồng thời tuyến tính theo từng đối số.

$\Omega$ được gọi là:
- **Đối xứng (symmetric):** nếu $\Omega(\mathbf{x}, \mathbf{y}) = \Omega(\mathbf{y}, \mathbf{x})$ với mọi $\mathbf{x}, \mathbf{y} \in V$
- **Xác định dương (positive definite):** nếu $\forall \mathbf{x} \in V \setminus \{\mathbf{0}\}: \Omega(\mathbf{x}, \mathbf{x}) > 0$ và $\Omega(\mathbf{0}, \mathbf{0}) = 0$

**Định nghĩa 3.3 — Tích Trong:** Ánh xạ song tuyến tính **đối xứng** và **xác định dương** $\Omega: V \times V \to \mathbb{R}$ được gọi là **tích trong** trên $V$, ký hiệu $\langle \mathbf{x}, \mathbf{y} \rangle$.

Cặp $(V, \langle \cdot, \cdot \rangle)$ gọi là **không gian tích trong** (inner product space). Khi dùng tích chấm (dot product) làm tích trong, ta gọi là **không gian vectơ Euclid**.

> **Tích chấm** là tích trong phổ biến nhất: $\langle \mathbf{x}, \mathbf{y} \rangle = \mathbf{x}^\top \mathbf{y} = \sum_{i=1}^n x_i y_i$. Nhưng không phải tích trong nào cũng là tích chấm — ví dụ với $V = \mathbb{R}^2$:
> $$\langle \mathbf{x}, \mathbf{y} \rangle := x_1 y_1 - (x_1 y_2 + x_2 y_1) + 2x_2 y_2 \tag{3.9}$$
> cũng là tích trong hợp lệ.

---

### Ma Trận Đối Xứng Xác Định Dương (Symmetric Positive Definite Matrices)

Cho không gian vectơ $n$ chiều $V$ với tích trong $\langle \cdot, \cdot \rangle$ và cơ sở $B = (\mathbf{b}_1, \ldots, \mathbf{b}_n)$. Viết $\mathbf{x} = \sum_i \psi_i \mathbf{b}_i$, $\mathbf{y} = \sum_j \lambda_j \mathbf{b}_j$, do tính song tuyến tính:

$$\langle \mathbf{x}, \mathbf{y} \rangle = \sum_{i=1}^n \sum_{j=1}^n \psi_i \langle \mathbf{b}_i, \mathbf{b}_j \rangle \lambda_j = \hat{\mathbf{x}}^\top A \hat{\mathbf{y}} \tag{3.10}$$

trong đó $A_{ij} := \langle \mathbf{b}_i, \mathbf{b}_j \rangle$ và $\hat{\mathbf{x}}, \hat{\mathbf{y}}$ là tọa độ theo cơ sở $B$. Tích trong được **xác định hoàn toàn** bởi ma trận $A$.

**Định nghĩa 3.4:** Ma trận đối xứng $A \in \mathbb{R}^{n \times n}$ thỏa mãn:
$$\forall \mathbf{x} \in V \setminus \{\mathbf{0}\}: \mathbf{x}^\top A \mathbf{x} > 0 \tag{3.11}$$
được gọi là **ma trận đối xứng xác định dương** (symmetric positive definite). Nếu chỉ có $\geq 0$ thì gọi là **nửa xác định dương** (positive semidefinite).

**Định lý 3.5:** $\langle \cdot, \cdot \rangle$ là tích trong $\iff$ tồn tại ma trận đối xứng xác định dương $A$ sao cho $\langle \mathbf{x}, \mathbf{y} \rangle = \hat{\mathbf{x}}^\top A \hat{\mathbf{y}}$.

### Ví dụ 3.4 — Kiểm tra xác định dương

Cho hai ma trận:
$$A_1 = \begin{bmatrix}9 & 6\\6 & 5\end{bmatrix}, \quad A_2 = \begin{bmatrix}9 & 6\\6 & 3\end{bmatrix} \tag{3.12}$$

**Kiểm tra $A_1$:**
$$\mathbf{x}^\top A_1 \mathbf{x} = 9x_1^2 + 12x_1x_2 + 5x_2^2 = (3x_1 + 2x_2)^2 + x_2^2 > 0 \tag{3.13}$$

Vì tổng hai bình phương, luôn dương với $\mathbf{x} \neq \mathbf{0}$ → $A_1$ **xác định dương** ✓

**Kiểm tra $A_2$:**
$$\mathbf{x}^\top A_2 \mathbf{x} = 9x_1^2 + 12x_1x_2 + 3x_2^2 = (3x_1 + 2x_2)^2 - x_2^2$$

Thử $\mathbf{x} = [2, -3]^\top$: $(6-6)^2 - 9 = -9 < 0$ → $A_2$ **không xác định dương** ✗

**Tính chất của ma trận xác định dương:**
- Nhân (null space) của $A$ chỉ chứa $\mathbf{0}$ (vì $\mathbf{x}^\top A \mathbf{x} > 0$ với $\mathbf{x} \neq \mathbf{0}$, nên $A\mathbf{x} \neq \mathbf{0}$)
- Tất cả phần tử trên đường chéo $a_{ii} > 0$ (vì $a_{ii} = \mathbf{e}_i^\top A \mathbf{e}_i > 0$)

## 3.3 Độ Dài và Khoảng Cách (Lengths and Distances)

### Lý thuyết

Tích trong **sinh ra** (induce) một chuẩn một cách tự nhiên:

$$\|\mathbf{x}\| := \sqrt{\langle \mathbf{x}, \mathbf{x} \rangle} \tag{3.16}$$

> **Lưu ý:** Không phải mọi chuẩn đều sinh từ tích trong — chuẩn Manhattan $\|\mathbf{x}\|_1 = \sum |x_i|$ là ví dụ chuẩn **không** có tích trong tương ứng.

**Bất đẳng thức Cauchy-Schwarz:**
$$|\langle \mathbf{x}, \mathbf{y} \rangle| \leq \|\mathbf{x}\| \cdot \|\mathbf{y}\| \tag{3.17}$$

**Định nghĩa 3.6 — Khoảng Cách và Metric:**

$$d(\mathbf{x}, \mathbf{y}) := \|\mathbf{x} - \mathbf{y}\| = \sqrt{\langle \mathbf{x} - \mathbf{y},\ \mathbf{x} - \mathbf{y} \rangle} \tag{3.21}$$

Khi dùng tích chấm, đây là **khoảng cách Euclid**. Ánh xạ $d: V \times V \to \mathbb{R}$ gọi là **metric**, thỏa mãn:

1. **Xác định dương:** $d(\mathbf{x}, \mathbf{y}) \geq 0$, và $d(\mathbf{x}, \mathbf{y}) = 0 \iff \mathbf{x} = \mathbf{y}$
2. **Đối xứng:** $d(\mathbf{x}, \mathbf{y}) = d(\mathbf{y}, \mathbf{x})$
3. **Bất đẳng thức tam giác:** $d(\mathbf{x}, \mathbf{z}) \leq d(\mathbf{x}, \mathbf{y}) + d(\mathbf{y}, \mathbf{z})$

> **Quan sát thú vị:** Tích trong và metric "chạy ngược chiều nhau" — $\mathbf{x}, \mathbf{y}$ giống nhau thì tích trong **lớn**, khoảng cách **nhỏ**.

### Ví dụ 3.5 — Độ dài phụ thuộc tích trong

Cho $\mathbf{x} = [1, 1]^\top \in \mathbb{R}^2$.

**Với tích chấm:** $\|\mathbf{x}\| = \sqrt{1^2 + 1^2} = \sqrt{2} \approx 1.414$

**Với tích trong** $\langle \mathbf{x}, \mathbf{y} \rangle = \mathbf{x}^\top \begin{bmatrix}1 & -\frac{1}{2}\\-\frac{1}{2} & 1\end{bmatrix} \mathbf{y}$:

$$\langle \mathbf{x}, \mathbf{x} \rangle = 1 - \frac{1}{2} - \frac{1}{2} + 1 = 1 \implies \|\mathbf{x}\| = 1 \tag{3.20}$$

Cùng một vectơ, tích trong khác nhau cho độ dài khác nhau.

## 3.4 Góc và Trực Giao (Angles and Orthogonality)

### Lý thuyết

Từ bất đẳng thức Cauchy-Schwarz: $-1 \leq \dfrac{\langle \mathbf{x}, \mathbf{y} \rangle}{\|\mathbf{x}\|\|\mathbf{y}\|} \leq 1$, nên tồn tại duy nhất $\omega \in [0, \pi]$ sao cho:

$$\cos \omega = \frac{\langle \mathbf{x}, \mathbf{y} \rangle}{\|\mathbf{x}\| \cdot \|\mathbf{y}\|} \tag{3.25}$$

$\omega$ là **góc** giữa hai vectơ $\mathbf{x}$ và $\mathbf{y}$.

**Định nghĩa 3.7 — Trực Giao:** Hai vectơ $\mathbf{x}, \mathbf{y}$ **trực giao** (orthogonal) nếu:
$$\langle \mathbf{x}, \mathbf{y} \rangle = 0 \quad (\text{ký hiệu } \mathbf{x} \perp \mathbf{y})$$

Nếu thêm $\|\mathbf{x}\| = 1 = \|\mathbf{y}\|$ thì gọi là **trực chuẩn** (orthonormal). Vectơ $\mathbf{0}$ trực giao với mọi vectơ.

> **Lưu ý quan trọng:** Tính trực giao **phụ thuộc vào tích trong được chọn** — hai vectơ có thể trực giao theo tích trong này nhưng không trực giao theo tích trong khác (xem Ví dụ 3.7 bên dưới).

### Ví dụ 3.6 — Tính góc giữa hai vectơ

Cho $\mathbf{x} = [1,1]^\top$, $\mathbf{y} = [1,2]^\top$, dùng tích chấm:

$$\cos\omega = \frac{\mathbf{x}^\top \mathbf{y}}{\sqrt{\mathbf{x}^\top\mathbf{x}} \cdot \sqrt{\mathbf{y}^\top\mathbf{y}}} = \frac{1\cdot1 + 1\cdot2}{\sqrt{2}\cdot\sqrt{5}} = \frac{3}{\sqrt{10}} \tag{3.26}$$

$$\omega = \arccos\!\left(\frac{3}{\sqrt{10}}\right) \approx 0.32 \text{ rad} \approx 18°$$

### Ví dụ 3.7 — Trực giao phụ thuộc tích trong

Cho $\mathbf{x} = [1,1]^\top$, $\mathbf{y} = [-1,1]^\top$.

**Với tích chấm:** $\langle \mathbf{x}, \mathbf{y} \rangle = -1 + 1 = 0$ → $\omega = 90°$, trực giao ✓

**Với tích trong** $\langle \mathbf{x}, \mathbf{y} \rangle = \mathbf{x}^\top \begin{bmatrix}2 & 0\\0 & 1\end{bmatrix} \mathbf{y}$:

$$\langle \mathbf{x}, \mathbf{y} \rangle = 2\cdot(1)(-1) + 1\cdot(1)(1) = -2 + 1 = -1$$

$$\|\mathbf{x}\| = \sqrt{2\cdot1 + 1} = \sqrt{3}, \quad \|\mathbf{y}\| = \sqrt{2\cdot1 + 1} = \sqrt{3}$$

$$\cos\omega = \frac{-1}{\sqrt{3}\cdot\sqrt{3}} = -\frac{1}{3} \implies \omega \approx 109.5° \tag{3.28}$$

Không trực giao theo tích trong này!

---

### Ma Trận Trực Giao (Orthogonal Matrix)

**Định nghĩa 3.8:** Ma trận vuông $A \in \mathbb{R}^{n \times n}$ là **ma trận trực giao** nếu các cột của nó **trực chuẩn** (orthonormal):

$$AA^\top = I = A^\top A \tag{3.29} \implies A^{-1} = A^\top \tag{3.30}$$

Tính chất đặc biệt — biến đổi bởi ma trận trực giao **bảo toàn độ dài và góc**:

**Bảo toàn độ dài:**
$$\|A\mathbf{x}\|^2 = (A\mathbf{x})^\top(A\mathbf{x}) = \mathbf{x}^\top A^\top A \mathbf{x} = \mathbf{x}^\top I \mathbf{x} = \|\mathbf{x}\|^2 \tag{3.31}$$

**Bảo toàn góc:**
$$\cos\omega = \frac{(A\mathbf{x})^\top(A\mathbf{y})}{\|A\mathbf{x}\|\|A\mathbf{y}\|} = \frac{\mathbf{x}^\top A^\top A \mathbf{y}}{\|\mathbf{x}\|\|\mathbf{y}\|} = \frac{\mathbf{x}^\top \mathbf{y}}{\|\mathbf{x}\|\|\mathbf{y}\|} \tag{3.32}$$

> Ma trận trực giao biểu diễn các **phép quay** (và có thể kết hợp phản chiếu) trong không gian.

---

## 3.5 Cơ Sở Trực Chuẩn (Orthonormal Basis)

### Lý thuyết

**Định nghĩa 3.9:** Cơ sở $\{\mathbf{b}_1, \ldots, \mathbf{b}_n\}$ của $V$ là **cơ sở trực chuẩn (ONB)** nếu:

$$\langle \mathbf{b}_i, \mathbf{b}_j \rangle = 0 \text{ với } i \neq j \quad \text{(trực giao)} \tag{3.33}$$
$$\langle \mathbf{b}_i, \mathbf{b}_i \rangle = 1 \quad \text{(chuẩn hóa)} \tag{3.34}$$

Nếu chỉ thỏa (3.33) thì gọi là **cơ sở trực giao** (orthogonal basis).

**Xây dựng ONB — Quy trình Gram-Schmidt:** Cho tập vectơ $\{\tilde{\mathbf{b}}_1, \ldots, \tilde{\mathbf{b}}_n\}$ chưa trực giao, chưa chuẩn hóa, xếp thành ma trận $\tilde{B}$, áp dụng khử Gauss trên ma trận bổ sung $[\tilde{B}\tilde{B}^\top \mid \tilde{B}]$ để thu được ONB.

### Ví dụ 3.8

Cơ sở chính tắc của $\mathbb{R}^n$ là ONB với tích chấm. Trong $\mathbb{R}^2$, cặp vectơ:

$$\mathbf{b}_1 = \frac{1}{\sqrt{2}}\begin{bmatrix}1\\1\end{bmatrix}, \quad \mathbf{b}_2 = \frac{1}{\sqrt{2}}\begin{bmatrix}1\\-1\end{bmatrix} \tag{3.35}$$

là ONB vì $\mathbf{b}_1^\top \mathbf{b}_2 = \frac{1}{2}(1-1) = 0$ và $\|\mathbf{b}_1\| = \|\mathbf{b}_2\| = 1$ ✓

---

## 3.6 Phần Bù Trực Giao (Orthogonal Complement)

### Lý thuyết

Cho không gian vectơ $V$ có số chiều $D$ và không gian con $U \subseteq V$ có số chiều $M$. **Phần bù trực giao** của $U$ là:

$$U^\perp := \{\mathbf{v} \in V : \langle \mathbf{u}, \mathbf{v} \rangle = 0,\ \forall \mathbf{u} \in U\}$$

$U^\perp$ là không gian con có số chiều $D - M$, và $U \cap U^\perp = \{\mathbf{0}\}$.

Mọi $\mathbf{x} \in V$ **phân tích duy nhất** thành:

$$\mathbf{x} = \underbrace{\sum_{m=1}^{M} \lambda_m \mathbf{b}_m}_{\in U} + \underbrace{\sum_{j=1}^{D-M} \psi_j \mathbf{b}_j^\perp}_{\in U^\perp} \tag{3.36}$$

> **Ứng dụng hình học:** Trong $\mathbb{R}^3$, mặt phẳng $U$ (2 chiều) có phần bù trực giao $U^\perp$ là đường thẳng (1 chiều) vuông góc với nó — chính là **vectơ pháp tuyến** (normal vector) của mặt phẳng.

---

## 3.7 Tích Trong của Hàm Số (Inner Product of Functions)

### Lý thuyết

Khái niệm tích trong có thể **mở rộng** từ vectơ hữu hạn chiều sang **hàm số**. Đối với hai hàm $u, v: \mathbb{R} \to \mathbb{R}$, tích trong được định nghĩa qua tích phân:

$$\langle u, v \rangle := \int_a^b u(x) v(x)\, dx \tag{3.37}$$

Tương tự với vectơ, nếu tích phân này bằng 0 thì $u$ và $v$ **trực giao**.

### Ví dụ 3.9 — $\sin$ và $\cos$ trực giao

Xét $u = \sin(x)$, $v = \cos(x)$, tích phân từ $-\pi$ đến $\pi$:

$$\langle \sin, \cos \rangle = \int_{-\pi}^{\pi} \sin(x)\cos(x)\, dx = \frac{1}{2}\int_{-\pi}^{\pi} \sin(2x)\, dx = 0$$

$f(x) = \sin(x)\cos(x)$ là hàm **lẻ** ($f(-x) = -f(x)$), nên tích phân trên đoạn đối xứng bằng 0 → $\sin \perp \cos$ ✓

> **Kết nối với Chuỗi Fourier:** Họ hàm $\{1, \cos(x), \cos(2x), \cos(3x), \ldots\}$ là **trực giao** trên $[-\pi, \pi]$. Chiếu một hàm lên không gian sinh bởi tập này chính là ý tưởng nền tảng của **chuỗi Fourier** — một ứng dụng trực tiếp của tích trong và trực giao trong phân tích hàm số.

## 3.8.2 Hình Chiếu lên Không Gian Con Tổng Quát

### Lý thuyết

Cho không gian con $U \subseteq \mathbb{R}^n$ với cơ sở $B = [\mathbf{b}_1, \ldots, \mathbf{b}_m]$. Hình chiếu $\pi_U(\mathbf{x})$ của vectơ $\mathbf{x}$ lên $U$ là vectơ $\mathbf{x}$ gần nhất trong $U$, tức:

$$\pi_U(\mathbf{x}) = B\boldsymbol{\lambda}$$

trong đó $\boldsymbol{\lambda}$ là nghiệm của **phương trình chuẩn** (normal equation):

$$B^\top B \boldsymbol{\lambda} = B^\top \mathbf{x} \implies \boldsymbol{\lambda} = (B^\top B)^{-1} B^\top \mathbf{x} \tag{3.57}$$

Từ đó:

$$\boxed{\pi_U(\mathbf{x}) = B(B^\top B)^{-1} B^\top \mathbf{x}} \tag{3.58}$$

$$\boxed{P_\pi = B(B^\top B)^{-1} B^\top} \tag{3.59}$$

> **Kết nối với trường hợp 1 chiều:** Khi $\dim(U) = 1$, $B^\top B \in \mathbb{R}$ là số vô hướng, và $P_\pi = \frac{BB^\top}{B^\top B}$ — đúng với công thức chiếu 1D.

> **Trường hợp cơ sở ONB:** Nếu $B$ là cơ sở trực chuẩn thì $B^\top B = I$, công thức đơn giản hóa thành:
> $$\pi_U(\mathbf{x}) = BB^\top \mathbf{x}, \quad \boldsymbol{\lambda} = B^\top \mathbf{x} \tag{3.65, 3.66}$$
> Không cần tính nghịch đảo — tiết kiệm tính toán đáng kể.

**Sai số hình chiếu (projection error / reconstruction error):**

$$\|\mathbf{x} - \pi_U(\mathbf{x})\|$$

---

### Ví dụ 3.11 — Hình chiếu lên không gian con 2 chiều trong $\mathbb{R}^3$

Cho $U = \text{span}\!\left[\begin{pmatrix}1\\1\\1\end{pmatrix}, \begin{pmatrix}0\\1\\2\end{pmatrix}\right] \subseteq \mathbb{R}^3$ và $\mathbf{x} = \begin{pmatrix}6\\0\\0\end{pmatrix}$.

**Bước 1:** Kiểm tra hai vectơ sinh độc lập tuyến tính → đúng. Xây dựng ma trận cơ sở:

$$B = \begin{pmatrix}1 & 0\\1 & 1\\1 & 2\end{pmatrix}$$

**Bước 2:** Tính $B^\top B$ và $B^\top \mathbf{x}$:

$$B^\top B = \begin{bmatrix}1 & 1 & 1\\0 & 1 & 2\end{bmatrix}\begin{pmatrix}1 & 0\\1 & 1\\1 & 2\end{pmatrix} = \begin{bmatrix}3 & 3\\3 & 5\end{bmatrix}$$

$$B^\top \mathbf{x} = \begin{bmatrix}1 & 1 & 1\\0 & 1 & 2\end{bmatrix}\begin{pmatrix}6\\0\\0\end{pmatrix} = \begin{bmatrix}6\\0\end{bmatrix}$$

**Bước 3:** Giải phương trình chuẩn $B^\top B \boldsymbol{\lambda} = B^\top \mathbf{x}$:

$$\begin{bmatrix}3 & 3\\3 & 5\end{bmatrix}\begin{bmatrix}\lambda_1\\\lambda_2\end{bmatrix} = \begin{bmatrix}6\\0\end{bmatrix} \tag{3.61}$$

$R_2 \leftarrow R_2 - R_1$:

$$\begin{bmatrix}3 & 3\\0 & 2\end{bmatrix}\begin{bmatrix}\lambda_1\\\lambda_2\end{bmatrix} = \begin{bmatrix}6\\-6\end{bmatrix} \implies \lambda_2 = -3,\ \lambda_1 = \frac{6 - 3(-3)}{3} = 5$$

$$\boldsymbol{\lambda} = \begin{bmatrix}5\\-3\end{bmatrix}$$

**Bước 4:** Tính hình chiếu:

$$\pi_U(\mathbf{x}) = B\boldsymbol{\lambda} = \begin{pmatrix}1 & 0\\1 & 1\\1 & 2\end{pmatrix}\begin{bmatrix}5\\-3\end{bmatrix} = \begin{pmatrix}5\\2\\-1\end{pmatrix} \tag{3.62}$$

**Bước 5:** Tính sai số hình chiếu:

$$\mathbf{x} - \pi_U(\mathbf{x}) = \begin{pmatrix}6\\0\\0\end{pmatrix} - \begin{pmatrix}5\\2\\-1\end{pmatrix} = \begin{pmatrix}1\\-2\\1\end{pmatrix}$$

$$\|\mathbf{x} - \pi_U(\mathbf{x})\| = \sqrt{1^2 + (-2)^2 + 1^2} = \sqrt{6} \tag{3.63}$$

**Bước 6:** Tính ma trận hình chiếu:

$$P_\pi = B(B^\top B)^{-1}B^\top = \frac{1}{6}\begin{pmatrix}5 & 2 & -1\\2 & 2 & 2\\-1 & 2 & 5\end{pmatrix} \tag{3.64}$$

**Kiểm tra:**
- Vectơ dư $\mathbf{x} - \pi_U(\mathbf{x}) = [1,-2,1]^\top$ trực giao với cả hai cột của $B$: $[1,1,1][1,-2,1]^\top = 0$ ✓ và $[0,1,2][1,-2,1]^\top = 0$ ✓
- $P_\pi^2 = P_\pi$ (ma trận hình chiếu là idempotent) ✓

> **Ứng dụng quan trọng:** Khi hệ $A\mathbf{x} = \mathbf{b}$ không có nghiệm (tức $\mathbf{b}$ không nằm trong không gian cột của $A$), ta tìm **nghiệm bình phương tối thiểu** bằng cách chiếu $\mathbf{b}$ lên không gian cột của $A$ — đây là nền tảng của hồi quy tuyến tính (Chương 9) và PCA (Chương 10).

## 3.8.3 Quy Trình Gram-Schmidt

### Lý thuyết

Cho cơ sở bất kỳ $(\mathbf{b}_1, \ldots, \mathbf{b}_n)$ của $V$. Quy trình **Gram-Schmidt** xây dựng cơ sở trực giao $(\mathbf{u}_1, \ldots, \mathbf{u}_n)$ theo công thức lặp:

$$\mathbf{u}_1 := \mathbf{b}_1 \tag{3.67}$$

$$\mathbf{u}_k := \mathbf{b}_k - \pi_{\text{span}[\mathbf{u}_1,\ldots,\mathbf{u}_{k-1}]}(\mathbf{b}_k), \quad k = 2, \ldots, n \tag{3.68}$$

**Ý tưởng:** Ở mỗi bước, lấy $\mathbf{b}_k$ và **trừ đi phần chiếu** của nó lên không gian đã xây dựng. Phần còn lại $\mathbf{u}_k$ vuông góc với tất cả $\mathbf{u}_1, \ldots, \mathbf{u}_{k-1}$. Chuẩn hóa $\mathbf{u}_k$ nếu muốn ONB.

```
Trực quan hóa quy trình:

Bước 1: u₁ = b₁  (giữ nguyên vectơ đầu tiên)

Bước 2: Chiếu b₂ lên span[u₁]
         Trừ phần chiếu → u₂ ⊥ u₁
         
         b₂ ──────────────────►
              ↘ π(b₂)          
         u₁ ──────► 
              ↑
              u₂ = b₂ - π(b₂)  (vuông góc với u₁)

Bước 3: Chiếu b₃ lên span[u₁, u₂]
         Trừ phần chiếu → u₃ ⊥ u₁, u₃ ⊥ u₂
```

### Ví dụ 3.12 — Gram-Schmidt trong $\mathbb{R}^2$

Cho cơ sở:

$$\mathbf{b}_1 = \begin{bmatrix}2\\0\end{bmatrix}, \quad \mathbf{b}_2 = \begin{bmatrix}1\\1\end{bmatrix} \tag{3.69}$$

**Bước 1:** $\mathbf{u}_1 := \mathbf{b}_1 = \begin{bmatrix}2\\0\end{bmatrix}$

**Bước 2:** Tính hình chiếu của $\mathbf{b}_2$ lên $\text{span}[\mathbf{u}_1]$:

$$\pi_{\text{span}[\mathbf{u}_1]}(\mathbf{b}_2) = \frac{\mathbf{u}_1 \mathbf{u}_1^\top}{\|\mathbf{u}_1\|^2}\mathbf{b}_2 = \frac{1}{4}\begin{bmatrix}2\\0\end{bmatrix}\begin{bmatrix}2 & 0\end{bmatrix}\begin{bmatrix}1\\1\end{bmatrix} = \frac{1}{4}\begin{bmatrix}4 & 0\\0 & 0\end{bmatrix}\begin{bmatrix}1\\1\end{bmatrix} = \begin{bmatrix}1\\0\end{bmatrix}$$

**Bước 3:** Tính $\mathbf{u}_2$:

$$\mathbf{u}_2 := \mathbf{b}_2 - \pi_{\text{span}[\mathbf{u}_1]}(\mathbf{b}_2) = \begin{bmatrix}1\\1\end{bmatrix} - \begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}0\\1\end{bmatrix} \tag{3.71}$$

**Kiểm tra:** $\mathbf{u}_1^\top \mathbf{u}_2 = [2,0][0,1]^\top = 0$ ✓

Nếu cần ONB, chuẩn hóa: $\hat{\mathbf{u}}_1 = \begin{bmatrix}1\\0\end{bmatrix}$, $\hat{\mathbf{u}}_2 = \begin{bmatrix}0\\1\end{bmatrix}$ — chính là cơ sở chính tắc.

## 3.8.4 Hình Chiếu lên Không Gian Con Affine

### Lý thuyết

Cho không gian affine $L = \mathbf{x}_0 + U$ với $\mathbf{b}_1, \mathbf{b}_2$ là cơ sở của $U$. Để chiếu $\mathbf{x}$ lên $L$, ta **quy về bài toán đã biết** (chiếu lên không gian con vectơ):

**Bước 1:** Dịch chuyển: trừ điểm hỗ trợ $\mathbf{x}_0$ khỏi cả $\mathbf{x}$ lẫn $L$:
$$L - \mathbf{x}_0 = U \quad \text{(không gian con vectơ)}$$

**Bước 2:** Chiếu $(\mathbf{x} - \mathbf{x}_0)$ lên $U$ bằng công thức (3.58).

**Bước 3:** Dịch ngược lại:

$$\boxed{\pi_L(\mathbf{x}) = \mathbf{x}_0 + \pi_U(\mathbf{x} - \mathbf{x}_0)} \tag{3.72}$$

Khoảng cách từ $\mathbf{x}$ đến $L$ bằng khoảng cách từ $(\mathbf{x} - \mathbf{x}_0)$ đến $U$:

$$d(\mathbf{x}, L) = \|\mathbf{x} - \pi_L(\mathbf{x})\| = d(\mathbf{x} - \mathbf{x}_0,\ U) \tag{3.73}$$

```
Trực quan:

    x
    |  \
    |   \  (dịch về gốc)
    |    \
    x₀───►  x - x₀
    |            |
    L = x₀ + U   U
    
    πL(x) = x₀ + πU(x - x₀)
```

# 3.9 Phép Quay (Rotations)

---

## 3.9.1 Phép Quay trong $\mathbb{R}^2$

### Lý thuyết

**Phép quay** là ánh xạ tuyến tính quay mặt phẳng một góc $\theta$ quanh gốc tọa độ. Theo quy ước: $\theta > 0$ là quay **ngược chiều kim đồng hồ**.

Từ lượng giác, ảnh của các vectơ cơ sở chính tắc qua phép quay góc $\theta$:

$$\Phi(\mathbf{e}_1) = \begin{bmatrix}\cos\theta\\\sin\theta\end{bmatrix}, \quad \Phi(\mathbf{e}_2) = \begin{bmatrix}-\sin\theta\\\cos\theta\end{bmatrix} \tag{3.75}$$

Do đó **ma trận quay** trong $\mathbb{R}^2$:

$$\boxed{R(\theta) = \begin{bmatrix}\cos\theta & -\sin\theta\\\sin\theta & \cos\theta\end{bmatrix}} \tag{3.76}$$

**Ví dụ số:** Quay $\mathbf{x} = [1, 0]^\top$ góc $\theta = 90°$:

$$R(90°)\begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}0 & -1\\1 & 0\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}0\\1\end{bmatrix}$$

Đúng như kỳ vọng: trục $x$ được quay thành trục $y$. ✓

---

## 3.9.2 Phép Quay trong $\mathbb{R}^3$

Trong $\mathbb{R}^3$, ta quay **một mặt phẳng** quanh **một trục**. Có 3 phép quay cơ bản quanh 3 trục tọa độ:

**Quay quanh trục $\mathbf{e}_1$** (mặt phẳng $e_2e_3$ quay):

$$R_1(\theta) = \begin{pmatrix}1 & 0 & 0\\0 & \cos\theta & -\sin\theta\\0 & \sin\theta & \cos\theta\end{pmatrix} \tag{3.77}$$

**Quay quanh trục $\mathbf{e}_2$** (mặt phẳng $e_1e_3$ quay):

$$R_2(\theta) = \begin{pmatrix}\cos\theta & 0 & \sin\theta\\0 & 1 & 0\\-\sin\theta & 0 & \cos\theta\end{pmatrix} \tag{3.78}$$

**Quay quanh trục $\mathbf{e}_3$** (mặt phẳng $e_1e_2$ quay):

$$R_3(\theta) = \begin{pmatrix}\cos\theta & -\sin\theta & 0\\\sin\theta & \cos\theta & 0\\0 & 0 & 1\end{pmatrix} \tag{3.79}$$

> **Cách nhớ:** Trục nào thì hàng và cột tương ứng của trục đó giữ nguyên (bằng 1), các chiều còn lại quay theo công thức $2 \times 2$ của $R(\theta)$.

**Ví dụ số:** Quay $\mathbf{x} = [1, 0, 0]^\top$ góc $90°$ quanh trục $\mathbf{e}_3$:

$$R_3(90°)\begin{pmatrix}1\\0\\0\end{pmatrix} = \begin{pmatrix}0 & -1 & 0\\1 & 0 & 0\\0 & 0 & 1\end{pmatrix}\begin{pmatrix}1\\0\\0\end{pmatrix} = \begin{pmatrix}0\\1\\0\end{pmatrix}$$

Trục $x$ được quay thành trục $y$ trong mặt phẳng $xy$ — đúng như kỳ vọng. ✓

---

## 3.9.3 Phép Quay Givens trong $\mathbb{R}^n$

### Lý thuyết

Tổng quát hóa lên $n$ chiều: **phép quay Givens** cố định $n-2$ chiều và quay trong một mặt phẳng 2D bất kỳ.

**Định nghĩa 3.11:** Ma trận quay Givens $R_{ij}(\theta) \in \mathbb{R}^{n \times n}$ là ma trận đơn vị $I_n$ với 4 phần tử được thay bởi:

$$r_{ii} = \cos\theta, \quad r_{ij} = -\sin\theta, \quad r_{ji} = \sin\theta, \quad r_{jj} = \cos\theta \tag{3.81}$$

Tức là:

$$R_{ij}(\theta) = \begin{pmatrix}I_{i-1} & & & & \\ & \cos\theta & \cdots & -\sin\theta & \\ & \vdots & I_{j-i-1} & \vdots & \\ & \sin\theta & \cdots & \cos\theta & \\ & & & & I_{n-j}\end{pmatrix} \tag{3.80}$$

Khi $n = 2$: thu được đúng $R(\theta)$ trong (3.76).

**Ứng dụng:** Phép quay Givens được dùng trong phân tích QR và các thuật toán số để "triệt tiêu" từng phần tử của ma trận.

---

## 3.9.4 Tính Chất của Phép Quay

| Tính chất | Phát biểu |
|---|---|
| **Bảo toàn khoảng cách** | $\|\mathbf{x} - \mathbf{y}\| = \|R_\theta\mathbf{x} - R_\theta\mathbf{y}\|$ |
| **Bảo toàn góc** | Góc giữa $R_\theta\mathbf{x}$ và $R_\theta\mathbf{y}$ = góc giữa $\mathbf{x}$ và $\mathbf{y}$ |
| **Không giao hoán trong $\mathbb{R}^{\geq 3}$** | $R_1 R_2 \neq R_2 R_1$ nói chung |
| **Giao hoán trong $\mathbb{R}^2$** | $R(\phi)R(\theta) = R(\theta)R(\phi)$ với mọi $\phi, \theta$ |
| **Tạo thành nhóm Abel** | Các phép quay trong $\mathbb{R}^2$ quanh cùng một điểm tạo thành nhóm Abel |

> **Tại sao $\mathbb{R}^3$ không giao hoán?** Thử tự kiểm tra: quay quyển sách 90° quanh trục dọc rồi 90° quanh trục ngang, so với đổi thứ tự — kết quả khác nhau. Đây là lý do thứ tự quay khớp nối robot rất quan trọng trong thực tế.

## Tổng Kết Mục 3.8 và 3.9

- **Hình chiếu trực giao** = tìm điểm gần nhất trong không gian con. Công thức tổng quát $P_\pi = B(B^\top B)^{-1}B^\top$; đơn giản hóa thành $BB^\top$ khi $B$ là ONB.
- **Gram-Schmidt** = cách xây dựng ONB từ bất kỳ cơ sở nào bằng cách lặp đi lặp lại: chiếu và trừ phần chiếu.
- **Chiếu lên không gian affine** = dịch về gốc → chiếu → dịch ngược lại.
- **Phép quay** = ánh xạ trực giao bảo toàn cả độ dài lẫn góc. Trong $\mathbb{R}^2$ giao hoán, trong $\mathbb{R}^{\geq 3}$ không giao hoán. Ma trận quay Givens tổng quát hóa lên $n$ chiều bằng cách quay từng mặt phẳng 2D.
