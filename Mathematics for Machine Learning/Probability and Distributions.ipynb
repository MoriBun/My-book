# Chương 6: Xác Suất và Phân Phối (Probability and Distributions)

Chương này xây dựng nền tảng xác suất cho machine learning — công cụ để **định lượng sự không chắc chắn** trong dữ liệu, mô hình, và dự đoán. Các khái niệm ở đây là nền tảng cho mô hình hóa xác suất (Chương 8), hồi quy (Chương 9), và ước lượng mật độ (Chương 11).

## 6.1 Xây Dựng Không Gian Xác Suất

### 6.1.1 Nền Tảng Triết Học

Logic Boolean cổ điển chỉ cho phép **đúng hoặc sai** — không đủ để mô tả lý luận plausible trong thực tế. Ví dụ: khi thấy bạn đến trễ, ta không thể kết luận chắc chắn lý do, nhưng có thể **đánh giá mức độ khả năng** của từng lý do.

Xác suất là **tổng quát hóa của logic Boolean** — thay vì đúng/sai, dùng **mức độ plausibility** là số thực liên tục.

**Ba tiêu chí Jaynes** cho hệ thống lý luận hợp lý:
1. Mức độ plausibility biểu diễn bằng **số thực**
2. Phải dựa trên **quy tắc lẽ thường**
3. Kết quả phải **nhất quán** theo 3 nghĩa: không mâu thuẫn, trung thực (dùng toàn bộ dữ liệu), và tái tạo được (cùng trạng thái hiểu biết → cùng giá trị)

**Định lý Cox–Jaynes:** Ba tiêu chí trên đủ để xác định duy nhất các quy tắc toán học của xác suất.

> **Hai trường phái diễn giải xác suất:**
>
> **Bayesian (chủ quan):** Xác suất = **mức độ tin tưởng** của người quan sát. Có thể dùng cho sự kiện xảy ra một lần duy nhất. Ví dụ: "Xác suất mô hình này đúng là 80%."
>
> **Frequentist (tần suất):** Xác suất = **tần suất tương đối** khi lặp lại thí nghiệm vô hạn lần. Ví dụ: "Xác suất đồng xu ngửa = 0.5."

---

### 6.1.2 Xác Suất và Biến Ngẫu Nhiên

#### Ba khái niệm cốt lõi của không gian xác suất (Kolmogorov)

**Không gian mẫu $\Omega$ (Sample Space):** Tập hợp tất cả kết quả có thể của thí nghiệm. Ví dụ: tung hai đồng xu → $\Omega = \{hh, ht, th, tt\}$.

**Không gian sự kiện $\mathcal{A}$ (Event Space):** Tập hợp các tập con của $\Omega$ mà ta có thể quan sát được. Với phân phối rời rạc, $\mathcal{A}$ thường là tập lũy thừa của $\Omega$.

**Xác suất $P$ (Probability):** Với mỗi sự kiện $A \in \mathcal{A}$, gán một số $P(A) \in [0,1]$ đo mức độ tin rằng sự kiện đó xảy ra. Yêu cầu: $P(\Omega) = 1$.

#### Biến Ngẫu Nhiên (Random Variable)

Thay vì làm việc trực tiếp với $\Omega$, ta dùng **biến ngẫu nhiên** $X: \Omega \to T$ — một **hàm** ánh xạ kết quả thí nghiệm sang **không gian đích** $T$ (target space) thuận tiện hơn.

> **Lưu ý quan trọng:** "Biến ngẫu nhiên" là tên gây nhầm lẫn — nó **không phải biến, mà là một hàm**. Tính ngẫu nhiên đến từ đầu vào $\omega \in \Omega$, không phải bản thân hàm.

Với tập con $S \subseteq T$:

$$P_X(S) = P(X \in S) = P(X^{-1}(S)) = P(\{\omega \in \Omega: X(\omega) \in S\}) \tag{6.8}$$

Hàm $P_X = P \circ X^{-1}$ gọi là **luật** hay **phân phối** của biến ngẫu nhiên $X$.

**Phân loại biến ngẫu nhiên:**
- $T$ hữu hạn hoặc đếm được → **biến ngẫu nhiên rời rạc** (Mục 6.2.1)
- $T = \mathbb{R}$ hoặc $T = \mathbb{R}^D$ → **biến ngẫu nhiên liên tục** (Mục 6.2.2)

### Ví dụ 6.1 — Rút đồng xu từ túi

**Thí nghiệm:** Rút 2 đồng xu (có hoàn lại) từ túi chứa xu $ (Mỹ) và xu £ (Anh). Xác suất rút được $ là 0.3.

**Không gian mẫu:** $\Omega = \{(\$,\$),\ (\$,£),\ (£,\$),\ (£,£)\}$

**Biến ngẫu nhiên** $X$ = số lần rút được $:

$$X((\$,\$)) = 2,\quad X((\$,£)) = 1,\quad X((£,\$)) = 1,\quad X((£,£)) = 0$$

**Không gian đích:** $T = \{0, 1, 2\}$

**Tính xác suất** (vì hai lần rút độc lập):

$$P(X=2) = P(\$)\cdot P(\$) = 0.3 \times 0.3 = 0.09 \tag{6.5}$$

$$P(X=1) = P(\$,£) + P(£,\$) = 0.3\times0.7 + 0.7\times0.3 = 0.42 \tag{6.6}$$

$$P(X=0) = P(£)\cdot P(£) = 0.7 \times 0.7 = 0.49 \tag{6.7}$$

**Kiểm tra:** $0.09 + 0.42 + 0.49 = 1$ ✓

### 6.1.3 Xác Suất và Thống Kê (Statistics)

Hai lĩnh vực liên quan nhưng khác nhau về **hướng suy luận**:

| | Xác Suất | Thống Kê |
|---|---|---|
| **Hướng** | Từ mô hình → dự đoán kết quả | Từ dữ liệu → suy luận mô hình |
| **Câu hỏi** | "Nếu mô hình là X, kết quả sẽ thế nào?" | "Dữ liệu này được sinh ra từ mô hình nào?" |
| **ML gần với** | Công cụ | **Mục tiêu** |

Machine learning về bản chất gần với **thống kê** — ta quan sát dữ liệu và cố gắng tìm mô hình giải thích dữ liệu đó tốt nhất. Xác suất là công cụ để làm điều này.

---

## Tổng Kết Mục 6.1

- **Không gian xác suất** = bộ ba $(\Omega, \mathcal{A}, P)$: không gian mẫu, không gian sự kiện, và xác suất.
- **Biến ngẫu nhiên** = hàm $X: \Omega \to T$ chuyển kết quả thô sang không gian số học thuận tiện. Phân phối của $X$ là $P_X = P \circ X^{-1}$.
- **Bayesian vs Frequentist** — hai cách diễn giải dẫn đến hai phương pháp khác nhau trong ML.
- **Xác suất vs Thống kê** — xác suất suy diễn xuôi (mô hình → kết quả), thống kê suy diễn ngược (kết quả → mô hình). ML dùng cả hai.
- **Mẹo thực tế:** Luôn kiểm tra biến ngẫu nhiên là **rời rạc hay liên tục** trước khi áp dụng công thức.

# 6.2 Xác Suất Rời Rạc và Liên Tục

---

## 6.2.1 Xác Suất Rời Rạc (Discrete Probabilities)

### Lý thuyết

Khi không gian đích $T$ rời rạc, ta mô tả phân phối qua **hàm khối xác suất** (probability mass function — pmf): $P(X = x)$ cho biết xác suất biến ngẫu nhiên $X$ nhận giá trị $x$.

Với **hai biến ngẫu nhiên** $X$ và $Y$, ba loại xác suất quan trọng:

**Xác suất đồng thời (Joint probability):**
$$P(X = x_i, Y = y_j) = \frac{n_{ij}}{N} \tag{6.9}$$

trong đó $n_{ij}$ là số sự kiện có trạng thái $x_i$ và $y_j$, $N$ là tổng số sự kiện.

**Xác suất biên (Marginal probability)** — tổng theo hàng hoặc cột:
$$P(X = x_i) = \frac{c_i}{N} = \frac{\sum_j n_{ij}}{N} \tag{6.10}$$
$$P(Y = y_j) = \frac{r_j}{N} = \frac{\sum_i n_{ij}}{N} \tag{6.11}$$

**Xác suất có điều kiện (Conditional probability):**
$$P(Y = y_j \mid X = x_i) = \frac{n_{ij}}{c_i} \tag{6.13}$$
$$P(X = x_i \mid Y = y_j) = \frac{n_{ij}}{r_j} \tag{6.14}$$

**Ràng buộc chuẩn hóa:**
$$\sum_i P(X = x_i) = 1, \quad \sum_j P(Y = y_j) = 1 \tag{6.12}$$

### Ví dụ 6.2 — Bảng xác suất đồng thời

Hình dung bảng xác suất với $X$ có 5 trạng thái và $Y$ có 3 trạng thái:

```
         x₁    x₂    x₃    x₄    x₅   │ rⱼ (tổng hàng)
    y₁ [ n₁₁  n₁₂  n₁₃  n₁₄  n₁₅ ]  │  r₁
    y₂ [ n₂₁  n₂₂  n₂₃  n₂₄  n₂₅ ]  │  r₂
    y₃ [ n₃₁  n₃₂  n₃₃  n₃₄  n₃₅ ]  │  r₃
    ─────────────────────────────────────────
    cᵢ   c₁    c₂    c₃    c₄    c₅  │  N
(tổng cột)
```

- **Xác suất biên** = tổng hàng / $N$ (cho $Y$) hoặc tổng cột / $N$ (cho $X$)
- **Xác suất đồng thời** = ô $n_{ij}$ / $N$
- **Xác suất có điều kiện** = ô $n_{ij}$ / tổng hàng (hoặc cột) tương ứng

> **Ứng dụng trong ML:** Phân phối rời rạc dùng để mô hình hóa **biến phân loại** (categorical variables) như nhãn lớp trong bài toán phân loại, hay ký tự trong nhận dạng chữ viết tay.

---

## 6.2.2 Xác Suất Liên Tục (Continuous Probabilities)

### Lý thuyết

Với biến ngẫu nhiên liên tục $X$ (giá trị trong $\mathbb{R}$ hoặc $\mathbb{R}^D$), không thể nói "xác suất tại một điểm" vì $P(X = x) = 0$ với mọi $x$ cụ thể. Thay vào đó, ta làm việc với **xác suất trên khoảng**.

**Định nghĩa 6.1 — Hàm mật độ xác suất (pdf):** Hàm $f: \mathbb{R}^D \to \mathbb{R}$ là pdf nếu:
1. $\forall \mathbf{x} \in \mathbb{R}^D: f(\mathbf{x}) \geq 0$
2. 
$$
\int_{\mathbb{R}^D} f(\mathbf{x})\, d\mathbf{x} = 1 \tag{6.15}
$$

Liên kết với biến ngẫu nhiên $X$:

$$P(a \leq X \leq b) = \int_a^b f(x)\, dx \tag{6.16}$$

> **Lưu ý quan trọng:** $f(x)$ có thể **lớn hơn 1** — đây không phải xác suất của một điểm, mà là **mật độ**. Chỉ cần tích phân trên toàn miền bằng 1.

**Định nghĩa 6.2 — Hàm phân phối tích lũy (cdf):** Với $\mathbf{X} = [X_1, \ldots, X_D]^\top$:

$$F_X(\mathbf{x}) = P(X_1 \leq x_1, \ldots, X_D \leq x_D) = \int_{-\infty}^{x_1} \cdots \int_{-\infty}^{x_D} f(z_1, \ldots, z_D)\, dz_1 \cdots dz_D \tag{6.17, 6.18}$$

---

## 6.2.3 So Sánh Rời Rạc và Liên Tục

### Ví dụ 6.3 — Phân phối đều

**Rời rạc:** $Z$ có 3 trạng thái $\{-1.1,\ 0.3,\ 1.5\}$, mỗi trạng thái xác suất $\frac{1}{3}$:

| $z$ | $-1.1$ | $0.3$ | $1.5$ |
|---|---|---|---|
| $P(Z=z)$ | $\frac{1}{3}$ | $\frac{1}{3}$ | $\frac{1}{3}$ |

Giá trị pmf $\leq 1$, tổng bằng 1.

**Liên tục:** $X$ đều trên $[0.9, 1.6]$, pdf:

$$p(x) = \frac{1}{1.6 - 0.9} = \frac{1}{0.7} \approx 1.43 > 1$$

$$\int_{0.9}^{1.6} p(x)\, dx = 1 \tag{6.19}$$

Giá trị pdf **có thể lớn hơn 1** — hoàn toàn hợp lệ vì pdf không phải xác suất điểm.

---

### Bảng Tóm Tắt Ký Hiệu (Bảng 6.1)

| Loại | "Xác suất điểm" | "Xác suất khoảng" |
|---|---|---|
| **Rời rạc** | $P(X=x)$ — hàm khối xác suất (pmf) | Không áp dụng |
| **Liên tục** | $p(x)$ — hàm mật độ xác suất (pdf) | $P(X \leq x)$ — hàm phân phối tích lũy (cdf) |

> **Lưu ý về ký hiệu trong ML:** Tài liệu ML thường dùng lẫn lộn "probability distribution" cho cả pmf và pdf. Theo ngữ cảnh:
> - Biến rời rạc: $P(X=x)$ là pmf, hay gọi tắt là "distribution"
> - Biến liên tục: $p(x)$ là pdf (density), $P(X \leq x)$ là cdf, cả hai đều hay bị gọi là "distribution"
>
> Khi đọc tài liệu, luôn kiểm tra **rời rạc hay liên tục** để hiểu đúng ý nghĩa.

---

## Tổng Kết Mục 6.2

- **Rời rạc** → pmf: $P(X=x) \in [0,1]$, tổng bằng 1. Xác suất đồng thời, biên, và có điều kiện đều đọc được từ bảng tần suất $n_{ij}/N$.
- **Liên tục** → pdf: $p(x) \geq 0$, tích phân bằng 1, **giá trị có thể > 1**. Xác suất chỉ có nghĩa trên **khoảng**, không phải tại một điểm.
- **cdf** = tích phân của pdf từ $-\infty$ đến $x$ — cách thống nhất để nói về xác suất tích lũy.
- Ba loại xác suất **đồng thời → biên → có điều kiện** là nền tảng cho quy tắc tổng và quy tắc tích ở Mục 6.3.

# 6.3 Quy Tắc Tổng, Quy Tắc Tích, và Định Lý Bayes

---

## Hai Quy Tắc Nền Tảng

Toàn bộ xác suất học máy xây dựng trên **đúng hai quy tắc** này.

---

### Quy Tắc Tổng (Sum Rule) — Marginalization

$$p(x) = \begin{cases} \sum_{y \in \mathcal{Y}} p(x, y) & \text{nếu } y \text{ rời rạc} \\ \int_\mathcal{Y} p(x, y)\, dy & \text{nếu } y \text{ liên tục} \end{cases} \tag{6.20}$$

Quy tắc tổng **lấy tích phân (hoặc tổng) ra** biến $y$ từ phân phối đồng thời để thu về phân phối biên của $x$. Vì vậy còn gọi là **tính chất biên hóa** (marginalization property).

Tổng quát hóa cho $\mathbf{x} = [x_1, \ldots, x_D]^\top$:

$$p(x_i) = \int p(x_1, \ldots, x_D)\, dx_{\backslash i} \tag{6.21}$$

trong đó $dx_{\backslash i}$ nghĩa là tích phân theo tất cả biến **trừ** $x_i$.

> **Thách thức tính toán:** Quy tắc tổng đòi hỏi tính tích phân (hoặc tổng) nhiều chiều — đây là một trong những bài toán khó nhất trong ML xác suất. Không có thuật toán thời gian đa thức tổng quát để tính chính xác.

---

### Quy Tắc Tích (Product Rule)

$$p(x, y) = p(y \mid x)\, p(x) \tag{6.22}$$

Mọi phân phối đồng thời đều phân tích được thành tích của **phân phối biên** và **phân phối có điều kiện**. Vì thứ tự tùy ý nên cũng có: $p(x, y) = p(x \mid y)\, p(y)$.

---

## Định Lý Bayes (Bayes' Theorem)

### Chứng minh từ quy tắc tích

Từ quy tắc tích theo hai thứ tự:

$$p(x, y) = p(x \mid y)\, p(y) \tag{6.24}$$
$$p(x, y) = p(y \mid x)\, p(x) \tag{6.25}$$

Vế phải bằng nhau nên:

$$p(x \mid y)\, p(y) = p(y \mid x)\, p(x) \implies \boxed{p(x \mid y) = \frac{p(y \mid x)\, p(x)}{p(y)}} \tag{6.26}$$

### Các thành phần của Bayes' Theorem

$$\underbrace{p(x \mid y)}_{\text{posterior}} = \frac{\overbrace{p(y \mid x)}^{\text{likelihood}}\ \overbrace{p(x)}^{\text{prior}}}{\underbrace{p(y)}_{\text{evidence}}} \tag{6.23}$$

**Prior $p(x)$:** Kiến thức **trước khi quan sát dữ liệu** về biến ẩn $x$. Ta chọn prior dựa trên hiểu biết chủ quan — nhưng prior phải có pdf/pmf khác 0 trên mọi giá trị $x$ hợp lý.

**Likelihood $p(y \mid x)$:** Mô tả **mối quan hệ giữa $x$ và $y$** — xác suất quan sát được dữ liệu $y$ nếu biết $x$. Lưu ý: likelihood **không phải** phân phối theo $x$, chỉ là phân phối theo $y$.

**Posterior $p(x \mid y)$:** Kiến thức **sau khi quan sát** $y$ — đây là đại lượng ta muốn tính trong thống kê Bayesian, kết hợp cả prior lẫn dữ liệu.

**Evidence (marginal likelihood) $p(y)$:**

$$p(y) = \int p(y \mid x)\, p(x)\, dx = \mathbb{E}_X[p(y \mid x)] \tag{6.27}$$

Evidence tích phân tử số của (6.23) theo $x$, nên **độc lập với $x$** và đóng vai trò **hằng số chuẩn hóa** đảm bảo posterior là phân phối hợp lệ. Ngoài chuẩn hóa, evidence còn đóng vai trò quan trọng trong **chọn mô hình Bayesian** (Mục 8.6) — nhưng tích phân này thường rất khó tính.

> **Cách đọc Bayes' Theorem:**
> $$\text{posterior} \propto \text{likelihood} \times \text{prior}$$
> Sau khi quan sát $y$, ta **cập nhật** niềm tin từ prior sang posterior bằng cách nhân với likelihood rồi chuẩn hóa.

> **Tại sao gọi là "probabilistic inverse"?** Likelihood $p(y \mid x)$ mô tả: biết $x$, dữ liệu $y$ như thế nào. Bayes' theorem **đảo ngược** quan hệ này: biết $y$, suy luận về $x$. Đây là lý do Bayes' theorem đôi khi được gọi là **nghịch đảo xác suất** (probabilistic inverse).

---

### Ví dụ trực quan — Chẩn đoán y tế

Cho bệnh hiếm gặp (1% dân số), test chẩn đoán có độ chính xác 99%:

- **Prior:** $p(\text{bệnh}) = 0.01$
- **Likelihood:** $p(\text{test}+ \mid \text{bệnh}) = 0.99$, $p(\text{test}+ \mid \text{không bệnh}) = 0.01$
- **Evidence:** $p(\text{test}+) = 0.99 \times 0.01 + 0.01 \times 0.99 = 0.0198$
- **Posterior:**

$$p(\text{bệnh} \mid \text{test}+) = \frac{0.99 \times 0.01}{0.0198} = 0.5$$

Dù test chính xác 99%, xác suất thực sự mắc bệnh khi test dương chỉ là **50%** — vì bệnh quá hiếm. Prior rất ảnh hưởng đến posterior!

## Tổng Kết Mục 6.3

| Quy tắc | Công thức | Ý nghĩa |
|---|---|---|
| **Sum rule** | $p(x) = \int p(x,y)\,dy$ | Lấy biên hóa — tích phân ra biến không quan tâm |
| **Product rule** | $p(x,y) = p(y\|x)p(x)$ | Phân tích đồng thời thành biên × có điều kiện |
| **Bayes' theorem** | $p(x\|y) = \frac{p(y\|x)p(x)}{p(y)}$ | Cập nhật niềm tin sau khi quan sát dữ liệu |

- **Hai quy tắc** (sum + product) là nền tảng của toàn bộ xác suất — mọi phép tính xác suất đều quy về hai quy tắc này.
- **Bayes' theorem** là hệ quả trực tiếp của product rule — cho phép **đảo ngược** quan hệ nhân quả từ $p(y\|x)$ sang $p(x\|y)$.
- **Evidence thường khó tính** vì đòi hỏi tích phân nhiều chiều → đây là động lực cho các phương pháp xấp xỉ (variational inference, MCMC) trong ML.

Khớp rồi! Phần bạn gửi lần này là đầu của 6.4 (Định nghĩa 6.3, 6.4 — kỳ vọng và trung bình), phần trước là phần tiếp theo (từ hiệp phương sai trở đi). Tôi đã có đủ toàn bộ 6.4. Đây là nội dung hoàn chỉnh:

# 6.4 Thống Kê Tóm Tắt và Độc Lập (Summary Statistics and Independence)

---

## 6.4.1 Trung Bình và Hiệp Phương Sai

### Kỳ Vọng (Expected Value)

**Định nghĩa 6.3:** Kỳ vọng của hàm $g: \mathbb{R} \to \mathbb{R}$ của biến ngẫu nhiên $X \sim p(x)$:

$$\mathbb{E}_X[g(x)] = \begin{cases} \int_\mathcal{X} g(x)p(x)\,dx & \text{nếu } X \text{ liên tục} \\ \sum_{x \in \mathcal{X}} g(x)p(x) & \text{nếu } X \text{ rời rạc} \end{cases} \tag{6.28, 6.29}$$

Với biến ngẫu nhiên **đa chiều** $X = [X_1, \ldots, X_D]^\top$, kỳ vọng tính từng chiều:

$$\mathbb{E}_X[g(\mathbf{x})] = \begin{pmatrix}\mathbb{E}_{X_1}[g(x_1)]\\\vdots\\\mathbb{E}_{X_D}[g(x_D)]\end{pmatrix} \in \mathbb{R}^D \tag{6.30}$$

> **Kỳ vọng là toán tử tuyến tính:** $\mathbb{E}_X[ag(x) + bh(x)] = a\mathbb{E}_X[g(x)] + b\mathbb{E}_X[h(x)]$ — tính chất này được dùng rất thường xuyên.

---

### Trung Bình (Mean)

**Định nghĩa 6.4:** Trung bình của biến ngẫu nhiên $X$ là trường hợp đặc biệt của kỳ vọng khi $g$ là hàm đồng nhất:

$$\mathbb{E}_X[\mathbf{x}] = \begin{pmatrix}\mathbb{E}_{X_1}[x_1]\\\vdots\\\mathbb{E}_{X_D}[x_D]\end{pmatrix} \in \mathbb{R}^D \tag{6.31}$$

trong đó $\mathbb{E}_{X_d}[x_d] = \int x_d p(x_d)\,dx_d$ (liên tục) hoặc $\sum_{x_i} x_i p(x_d = x_i)$ (rời rạc).

**Ba đại lượng đo "trung tâm" của phân phối:**

| Đại lượng | Định nghĩa | Đặc điểm |
|---|---|---|
| **Mean** | $\mathbb{E}[x]$ | Nhạy cảm với outlier |
| **Median** | Giá trị chia đôi phân phối (cdf = 0.5) | Bền vững hơn với outlier, tốt cho phân phối lệch |
| **Mode** | Giá trị có xác suất/mật độ cao nhất | Có thể có nhiều mode (multimodal) |

> **Lưu ý:** Trong không gian nhiều chiều, median khó tổng quát hóa vì không có cách tự nhiên để sắp xếp thứ tự các điểm đa chiều.

---

### Hiệp Phương Sai (Covariance)

**Định nghĩa 6.5 — Hiệp phương sai một chiều:**

$$\text{Cov}_{X,Y}[x,y] := \mathbb{E}_{X,Y}[(x - \mathbb{E}[x])(y - \mathbb{E}[y])] \tag{6.35}$$

Hay viết gọn hơn:

$$\text{Cov}[x,y] = \mathbb{E}[xy] - \mathbb{E}[x]\mathbb{E}[y] \tag{6.36}$$

> Công thức (6.36) dễ nhớ hơn: **"trung bình của tích trừ tích của trung bình"**.

**Định nghĩa 6.6 — Hiệp phương sai đa chiều:** Với $X \in \mathbb{R}^D$, $Y \in \mathbb{R}^E$:

$$\text{Cov}[\mathbf{x}, \mathbf{y}] = \mathbb{E}[\mathbf{x}\mathbf{y}^\top] - \mathbb{E}[\mathbf{x}]\mathbb{E}[\mathbf{y}]^\top = \text{Cov}[\mathbf{y}, \mathbf{x}]^\top \in \mathbb{R}^{D \times E} \tag{6.37}$$

**Định nghĩa 6.7 — Phương sai và ma trận hiệp phương sai:** Với $X \in \mathbb{R}^D$, $\boldsymbol{\mu} = \mathbb{E}[\mathbf{x}]$:

$$\mathbb{V}_X[\mathbf{x}] = \text{Cov}[\mathbf{x}, \mathbf{x}] = \mathbb{E}[(\mathbf{x}-\boldsymbol{\mu})(\mathbf{x}-\boldsymbol{\mu})^\top] = \begin{pmatrix}\text{Cov}[x_1,x_1] & \cdots & \text{Cov}[x_1,x_D]\\\vdots & \ddots & \vdots\\\text{Cov}[x_D,x_1] & \cdots & \text{Cov}[x_D,x_D]\end{pmatrix} \tag{6.38}$$

**Đường chéo** = phương sai của từng chiều biên $p(x_i)$. **Ngoài đường chéo** = cross-covariance $\text{Cov}[x_i, x_j]$.

Ma trận hiệp phương sai là **đối xứng và nửa xác định dương**.

**Phương sai một chiều** $\mathbb{V}[x] = \text{Cov}[x,x]$, căn bậc hai là **độ lệch chuẩn** $\sigma(x)$.

**Định nghĩa 6.8 — Tương quan (Correlation):**

$$\text{corr}[x,y] = \frac{\text{Cov}[x,y]}{\sqrt{\mathbb{V}[x]\mathbb{V}[y]}} \in [-1, 1] \tag{6.40}$$

Tương quan là hiệp phương sai đã **chuẩn hóa** — loại bỏ ảnh hưởng của đơn vị đo.
- $\text{corr} > 0$: $x$ tăng thì $y$ tăng
- $\text{corr} < 0$: $x$ tăng thì $y$ giảm
- $\text{corr} = 0$: không tương quan tuyến tính

---

## 6.4.2 Trung Bình và Hiệp Phương Sai Thực Nghiệm

Trong ML, ta ước lượng từ dữ liệu hữu hạn $\mathbf{x}_1, \ldots, \mathbf{x}_N$.

**Định nghĩa 6.9:**

$$\bar{\mathbf{x}} := \frac{1}{N}\sum_{n=1}^N \mathbf{x}_n \tag{6.41}$$

$$\Sigma := \frac{1}{N}\sum_{n=1}^N (\mathbf{x}_n - \bar{\mathbf{x}})(\mathbf{x}_n - \bar{\mathbf{x}})^\top \tag{6.42}$$

> **Biased vs Unbiased:** Công thức (6.42) dùng $\frac{1}{N}$ là ước lượng **có thiên lệch** (biased). Ước lượng không thiên lệch dùng $\frac{1}{N-1}$.

---

## 6.4.3 Ba Cách Biểu Diễn Phương Sai

**Cách 1 — Định nghĩa chuẩn:**
$$\mathbb{V}[x] := \mathbb{E}[(x - \mu)^2] \tag{6.43}$$
Cần **hai lần duyệt** dữ liệu: lần 1 tính $\mu$, lần 2 tính phương sai.

**Cách 2 — Raw-score formula (một lần duyệt):**
$$\mathbb{V}[x] = \mathbb{E}[x^2] - (\mathbb{E}[x])^2 \tag{6.44}$$
Nhớ bằng: **"trung bình của bình phương trừ bình phương của trung bình"**. Tích lũy $x_i$ và $x_i^2$ đồng thời trong một lần duyệt — nhưng có thể **mất ổn định số** khi hai số hạng lớn và xấp xỉ nhau.

**Cách 3 — Tổng khoảng cách từng cặp:**
$$\frac{1}{N^2}\sum_{i,j=1}^N (x_i - x_j)^2 = 2\left(\frac{1}{N}\sum_i x_i^2 - \left(\frac{1}{N}\sum_i x_i\right)^2\right) \tag{6.45}$$

Tổng $N^2$ khoảng cách từng cặp = 2 lần raw-score formula. Ý nghĩa: **khoảng cách từng cặp tương đương khoảng cách từ tâm** — tính $N$ số hạng (mean + variance) thay vì $N^2$.

---

## 6.4.4 Tổng và Biến Đổi Biến Ngẫu Nhiên

Với $X, Y$ có trạng thái $\mathbf{x}, \mathbf{y} \in \mathbb{R}^D$:

$$\mathbb{E}[\mathbf{x} + \mathbf{y}] = \mathbb{E}[\mathbf{x}] + \mathbb{E}[\mathbf{y}] \tag{6.46}$$
$$\mathbb{V}[\mathbf{x} + \mathbf{y}] = \mathbb{V}[\mathbf{x}] + \mathbb{V}[\mathbf{y}] + \text{Cov}[\mathbf{x},\mathbf{y}] + \text{Cov}[\mathbf{y},\mathbf{x}] \tag{6.48}$$

**Biến đổi affine** $\mathbf{y} = A\mathbf{x} + \mathbf{b}$ — kết quả cực kỳ quan trọng:

$$\mathbb{E}_Y[\mathbf{y}] = A\boldsymbol{\mu} + \mathbf{b} \tag{6.50}$$
$$\mathbb{V}_Y[\mathbf{y}] = A\Sigma A^\top \tag{6.51}$$
$$\text{Cov}[\mathbf{x}, \mathbf{y}] = \Sigma A^\top \tag{6.52}$$

> **Ứng dụng quan trọng:** Khi dữ liệu $\mathbf{x} \sim \mathcal{N}(\boldsymbol{\mu}, \Sigma)$ và ta áp dụng biến đổi tuyến tính $A$, kết quả vẫn là Gaussian với mean và covariance tính theo (6.50)–(6.51). Đây là nền tảng của nhiều thuật toán ML.

---

## 6.4.5 Độc Lập Thống Kê

**Định nghĩa 6.10 — Độc lập thống kê:**

$$X, Y \text{ độc lập} \iff p(x,y) = p(x)p(y) \tag{6.53}$$

Nếu $X, Y$ độc lập thì:
- $p(y \mid x) = p(y)$ và $p(x \mid y) = p(x)$
- $\mathbb{V}[x + y] = \mathbb{V}[x] + \mathbb{V}[y]$
- $\text{Cov}[x, y] = 0$

> **Lưu ý quan trọng:** Chiều ngược lại **không đúng**! $\text{Cov}[x,y] = 0$ **không** suy ra $X, Y$ độc lập — covariance chỉ đo **phụ thuộc tuyến tính**, không phát hiện được phụ thuộc phi tuyến.

**Ví dụ 6.5:** Cho $X$ có mean 0 và $\mathbb{E}[x^3] = 0$, đặt $Y = x^2$ (rõ ràng phụ thuộc $X$):

$$\text{Cov}[x, y] = \mathbb{E}[xy] - \mathbb{E}[x]\mathbb{E}[y] = \mathbb{E}[x^3] - 0 = 0 \tag{6.54}$$

Covariance = 0 nhưng $Y$ hoàn toàn phụ thuộc $X$!

**i.i.d. (independent and identically distributed):** Giả thiết phổ biến nhất trong ML — $X_1, \ldots, X_N$ độc lập và cùng phân phối.

**Định nghĩa 6.11 — Độc lập có điều kiện:**

$$X \perp\!\!\!\perp Y \mid Z \iff p(x,y \mid z) = p(x \mid z)\, p(y \mid z) \quad \forall z \in \mathcal{Z} \tag{6.55}$$

Tương đương: $p(x \mid y, z) = p(x \mid z)$ — biết $z$ rồi thì thêm $y$ không cung cấp thông tin về $x$.

---

## 6.4.6 Tích Trong của Biến Ngẫu Nhiên

Với biến ngẫu nhiên mean 0, định nghĩa tích trong:

$$\langle X, Y \rangle := \text{Cov}[x, y] \tag{6.59}$$

Từ đó:
- **Độ dài** của biến ngẫu nhiên: $\|X\| = \sqrt{\text{Cov}[x,x]} = \sigma[x]$ — độ lệch chuẩn
- **Góc** giữa hai biến ngẫu nhiên:

$$\cos\theta = \frac{\langle X, Y\rangle}{\|X\|\|Y\|} = \frac{\text{Cov}[x,y]}{\sqrt{\mathbb{V}[x]\mathbb{V}[y]}} = \text{corr}[x,y] \tag{6.61}$$

> **Kết nối đẹp:** **Tương quan = cosin của góc** giữa hai biến ngẫu nhiên khi xem chúng là vectơ trong không gian tích trong. Hai biến **không tương quan** ($\text{Cov} = 0$) tương ứng với hai vectơ **trực giao** ($\cos\theta = 0$). Định lý Pythagoras $\mathbb{V}[x+y] = \mathbb{V}[x] + \mathbb{V}[y]$ chính là hệ quả của tính trực giao này.

## Tổng Kết Mục 6.4

| Đại lượng | Công thức | Ý nghĩa |
|---|---|---|
| Mean | $\mathbb{E}[\mathbf{x}]$ | Vị trí trung tâm |
| Variance | $\mathbb{E}[(x-\mu)^2]$ | Độ phân tán |
| Covariance | $\mathbb{E}[xy] - \mathbb{E}[x]\mathbb{E}[y]$ | Phụ thuộc tuyến tính |
| Correlation | $\text{Cov}/(\sigma_x\sigma_y)$ | Phụ thuộc tuyến tính chuẩn hóa |
| Independence | $p(x,y) = p(x)p(y)$ | Không có phụ thuộc (kể cả phi tuyến) |

- **Biến đổi affine** $\mathbf{y} = A\mathbf{x}+\mathbf{b}$: mean biến đổi tuyến tính, covariance biến đổi thành $A\Sigma A^\top$ — công thức nền tảng cho Gaussian và nhiều mô hình ML.
- **Cov = 0 không suy ra độc lập** — chỉ nắm được phụ thuộc tuyến tính.
- **Tương quan = cosin góc** — cầu nối giữa hình học (Chương 3) và xác suất.

# 6.5 Phân Phối Gaussian

Phân phối Gaussian (hay phân phối chuẩn) là phân phối **quan trọng nhất** trong ML — xuất hiện tự nhiên qua **Định lý Giới hạn Trung tâm** (Central Limit Theorem): tổng của nhiều biến ngẫu nhiên i.i.d. hội tụ về Gaussian.

---

## Định Nghĩa

**Một chiều (univariate):**

$$p(x \mid \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right) \tag{6.62}$$

**Đa chiều (multivariate):** Với $\mathbf{x} \in \mathbb{R}^D$:

$$p(\mathbf{x} \mid \boldsymbol{\mu}, \Sigma) = (2\pi)^{-D/2} |\Sigma|^{-1/2} \exp\!\left(-\frac{1}{2}(\mathbf{x}-\boldsymbol{\mu})^\top \Sigma^{-1} (\mathbf{x}-\boldsymbol{\mu})\right) \tag{6.63}$$

Ký hiệu: $p(\mathbf{x}) = \mathcal{N}(\mathbf{x} \mid \boldsymbol{\mu}, \Sigma)$ hoặc $X \sim \mathcal{N}(\boldsymbol{\mu}, \Sigma)$.

**Phân phối chuẩn tắc (standard normal):** $\boldsymbol{\mu} = \mathbf{0}$, $\Sigma = I$.

> **Tại sao Gaussian quan trọng trong ML?**
> - Có biểu thức **dạng đóng** (closed-form) cho phân phối biên và có điều kiện
> - Biến đổi tuyến tính của Gaussian **vẫn là Gaussian** — không cần đổi biến phức tạp
> - Được đặc trưng hoàn toàn bởi **mean và covariance** — đơn giản để làm việc
> - Ứng dụng rộng: hồi quy tuyến tính (Ch.9), Kalman filter, Gaussian process, VAE,...

---

## 6.5.1 Biên và Có Điều Kiện của Gaussian đều là Gaussian

Cho phân phối đồng thời Gaussian:

$$p(\mathbf{x}, \mathbf{y}) = \mathcal{N}\!\left(\begin{bmatrix}\boldsymbol{\mu}_x\\\boldsymbol{\mu}_y\end{bmatrix}, \begin{bmatrix}\Sigma_{xx} & \Sigma_{xy}\\\Sigma_{yx} & \Sigma_{yy}\end{bmatrix}\right) \tag{6.64}$$

**Phân phối biên** (áp dụng sum rule, tích phân ra $\mathbf{y}$):

$$p(\mathbf{x}) = \mathcal{N}(\mathbf{x} \mid \boldsymbol{\mu}_x, \Sigma_{xx}) \tag{6.68}$$

Chỉ đơn giản là lấy phần mean và covariance tương ứng — **bỏ qua** (tích phân ra) mọi thứ không quan tâm.

**Phân phối có điều kiện** (khi quan sát $\mathbf{y}$):

$$p(\mathbf{x} \mid \mathbf{y}) = \mathcal{N}(\mathbf{x} \mid \boldsymbol{\mu}_{x|y}, \Sigma_{x|y}) \tag{6.65}$$

$$\boldsymbol{\mu}_{x|y} = \boldsymbol{\mu}_x + \Sigma_{xy}\Sigma_{yy}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y) \tag{6.66}$$

$$\Sigma_{x|y} = \Sigma_{xx} - \Sigma_{xy}\Sigma_{yy}^{-1}\Sigma_{yx} \tag{6.67}$$

> **Cách nhớ công thức có điều kiện:**
> - Mean mới = mean cũ + **hiệu chỉnh** dựa trên khoảng cách $(\mathbf{y} - \boldsymbol{\mu}_y)$ nhân với "độ tương quan" $\Sigma_{xy}\Sigma_{yy}^{-1}$
> - Covariance mới = covariance cũ **trừ đi** phần thông tin thu được từ quan sát $\mathbf{y}$ — luôn nhỏ hơn hoặc bằng $\Sigma_{xx}$, tức **quan sát làm giảm bất định**

> **Ứng dụng:**
> - **Kalman filter:** cập nhật trạng thái hệ thống khi nhận quan sát mới — chính là tính Gaussian conditional
> - **Gaussian Process:** posterior trên hàm khi điều kiện hóa theo dữ liệu quan sát
> - **PPCA (Mục 10.7):** mô hình tuyến tính ẩn Gaussian

### Ví dụ 6.6 — Tính toán tường minh

Cho phân phối đồng thời:

$$p(x_1, x_2) = \mathcal{N}\!\left(\begin{bmatrix}0\\2\end{bmatrix}, \begin{bmatrix}0.3 & -1\\-1 & 5\end{bmatrix}\right) \tag{6.69}$$

**Phân phối biên** — đọc trực tiếp:
$$p(x_1) = \mathcal{N}(0,\ 0.3) \tag{6.73}$$

**Phân phối có điều kiện** khi $x_2 = -1$: Áp dụng (6.66) và (6.67):

$$\mu_{x_1|x_2=-1} = 0 + (-1) \cdot \frac{1}{5} \cdot (-1 - 2) = 0 + (-1)(0.2)(-3) = 0.6 \tag{6.70}$$

$$\sigma^2_{x_1|x_2=-1} = 0.3 - (-1) \cdot \frac{1}{5} \cdot (-1) = 0.3 - 0.2 = 0.1 \tag{6.71}$$

$$p(x_1 \mid x_2 = -1) = \mathcal{N}(0.6,\ 0.1) \tag{6.72}$$

**Nhận xét:** Mean dịch từ 0 lên 0.6 (do $x_1$ và $x_2$ tương quan âm, quan sát $x_2 < \mu_{x_2}$ kéo $x_1$ lên). Variance giảm từ 0.3 xuống 0.1 — quan sát $x_2$ giảm bất định về $x_1$.

---

## 6.5.2 Tích của Hai Gaussian

$$\mathcal{N}(\mathbf{x} \mid \mathbf{a}, A) \cdot \mathcal{N}(\mathbf{x} \mid \mathbf{b}, B) = c \cdot \mathcal{N}(\mathbf{x} \mid \mathbf{c}, C) \tag{6.74-6.76}$$

trong đó:

$$C = (A^{-1} + B^{-1})^{-1} \tag{6.74}$$

$$\mathbf{c} = C(A^{-1}\mathbf{a} + B^{-1}\mathbf{b}) \tag{6.75}$$

$$c = \mathcal{N}(\mathbf{a} \mid \mathbf{b},\ A + B) = \mathcal{N}(\mathbf{b} \mid \mathbf{a},\ A + B) \tag{6.76}$$

> **Ứng dụng:** Trong Bayesian linear regression (Ch.9), posterior = likelihood × prior, tức tích của hai Gaussian → vẫn là Gaussian (có scaling constant $c$). Đây là lý do Gaussian là **prior liên hợp** (conjugate prior) của chính nó.

---

## 6.5.3 Tổng và Biến Đổi Tuyến Tính

**Tổng hai Gaussian độc lập:**

$$p(\mathbf{x} + \mathbf{y}) = \mathcal{N}(\boldsymbol{\mu}_x + \boldsymbol{\mu}_y,\ \Sigma_x + \Sigma_y) \tag{6.78}$$

**Tổng có trọng số:**

$$p(a\mathbf{x} + b\mathbf{y}) = \mathcal{N}(a\boldsymbol{\mu}_x + b\boldsymbol{\mu}_y,\ a^2\Sigma_x + b^2\Sigma_y) \tag{6.79}$$

**Biến đổi tuyến tính** $\mathbf{y} = A\mathbf{x}$ với $X \sim \mathcal{N}(\boldsymbol{\mu}, \Sigma)$:

$$p(\mathbf{y}) = \mathcal{N}(\mathbf{y} \mid A\boldsymbol{\mu},\ A\Sigma A^\top) \tag{6.88}$$

> **Tính chất vàng của Gaussian:** Mọi biến đổi **affine** của Gaussian đều **vẫn là Gaussian** — chỉ cần biến đổi mean và covariance theo công thức tuyến tính. Không cần tính tích phân phức tạp.

**Biến đổi ngược:** Khi biết $p(\mathbf{y}) = \mathcal{N}(\mathbf{y} \mid A\mathbf{x}, \Sigma)$ với $A \in \mathbb{R}^{M \times N}$, $M \geq N$, dùng giả nghịch đảo:

$$\mathbf{y} = A\mathbf{x} \iff (A^\top A)^{-1}A^\top \mathbf{y} = \mathbf{x} \tag{6.90}$$

$$p(\mathbf{x}) = \mathcal{N}\!\left(\mathbf{x} \mid (A^\top A)^{-1}A^\top\mathbf{y},\ (A^\top A)^{-1}A^\top\Sigma A(A^\top A)^{-1}\right) \tag{6.91}$$

### Định lý 6.12 — Trung bình và Phương sai của Mixture Gaussian

Cho mixture: $p(x) = \alpha p_1(x) + (1-\alpha)p_2(x)$ với $0 < \alpha < 1$:

**Mean:**

$$\mathbb{E}[x] = \alpha\mu_1 + (1-\alpha)\mu_2 \tag{6.81}$$

**Variance** (dùng raw-score formula $\mathbb{V}[x] = \mathbb{E}[x^2] - (\mathbb{E}[x])^2$):

$$\mathbb{V}[x] = \underbrace{\left[\alpha\sigma_1^2 + (1-\alpha)\sigma_2^2\right]}_{\text{trung bình có trọng số của variance}} + \underbrace{\left[\alpha\mu_1^2 + (1-\alpha)\mu_2^2\right] - \left[\alpha\mu_1 + (1-\alpha)\mu_2\right]^2}_{\text{variance của mean}} \tag{6.82}$$

> **Ý nghĩa:** Phương sai tổng = **trung bình có trọng số của variance từng thành phần** + **variance của mean các thành phần**. Đây là trường hợp cụ thể của **Law of Total Variance**: $\mathbb{V}[X] = \mathbb{E}[\mathbb{V}[X|Y]] + \mathbb{V}[\mathbb{E}[X|Y]]$.

---

## 6.5.4 Lấy Mẫu từ Phân Phối Gaussian Đa Chiều

Để lấy mẫu từ $\mathcal{N}(\boldsymbol{\mu}, \Sigma)$, thực hiện 3 bước:

**Bước 1:** Lấy mẫu $\mathbf{z} \sim \mathcal{N}(\mathbf{0}, I)$ từ phân phối chuẩn tắc (dùng Box-Müller transform từ uniform $[0,1]$).

**Bước 2:** Phân tích Cholesky: $\Sigma = AA^\top$ (với $A$ là ma trận tam giác dưới).

**Bước 3:** Biến đổi: $\mathbf{x} = A\mathbf{z} + \boldsymbol{\mu}$

Vì $\mathbf{z} \sim \mathcal{N}(\mathbf{0}, I)$ nên $A\mathbf{z} + \boldsymbol{\mu} \sim \mathcal{N}(\boldsymbol{\mu}, AA^\top) = \mathcal{N}(\boldsymbol{\mu}, \Sigma)$ ✓

> **Kết nối với Cholesky (Mục 4.3):** Đây là ứng dụng trực tiếp của phân tích Cholesky — biến đổi mẫu từ phân phối đơn giản sang phân phối phức tạp hơn. Kỹ thuật này cũng là nền tảng của **reparametrization trick** trong VAE.

---

## Tổng Kết Mục 6.5

| Tính chất | Nội dung |
|---|---|
| **Định nghĩa** | $\mathcal{N}(\boldsymbol{\mu}, \Sigma)$, đặc trưng hoàn toàn bởi mean và covariance |
| **Phân phối biên** | Vẫn là Gaussian, đọc trực tiếp mean/covariance tương ứng |
| **Phân phối có điều kiện** | Vẫn là Gaussian, mean được cập nhật theo quan sát |
| **Biến đổi affine** | Vẫn là Gaussian: $\mathbf{y}=A\mathbf{x}+\mathbf{b} \Rightarrow \mathcal{N}(A\boldsymbol{\mu}+\mathbf{b},\ A\Sigma A^\top)$ |
| **Tích hai Gaussian** | Gaussian × scaling constant |
| **Lấy mẫu** | Cholesky + biến đổi tuyến tính từ $\mathcal{N}(\mathbf{0},I)$ |

Gaussian là phân phối **đóng** dưới mọi phép toán ML quan trọng — đây là lý do nó được dùng ở khắp nơi trong Chương 9, 10, 11.

# 6.6 Liên Hợp và Họ Hàm Mũ (Conjugacy and the Exponential Family)

---

## Ba Tiêu Chí cho Phân Phối Tốt trong ML

1. **Tính đóng (Closure):** Áp dụng quy tắc xác suất (Bayes' theorem) trả về cùng loại phân phối
2. **Số tham số hữu hạn:** Thu thêm dữ liệu không cần thêm tham số
3. **Ước lượng tham số thuận tiện:** Học từ dữ liệu có tính chất tốt về mặt tính toán

Họ **exponential family** cung cấp sự cân bằng đúng đắn giữa tính tổng quát và thuận tiện tính toán.

---

## Ba Phân Phối "Có Tên" Quan Trọng

### Phân Phối Bernoulli

Mô hình biến nhị phân $X \in \{0,1\}$ với tham số $\mu \in [0,1]$:

$$p(x \mid \mu) = \mu^x(1-\mu)^{1-x} \tag{6.92}$$

$$\mathbb{E}[x] = \mu, \quad \mathbb{V}[x] = \mu(1-\mu) \tag{6.93, 6.94}$$

Ứng dụng: mô hình hóa xác suất "mặt ngửa" khi tung đồng xu.

> **Thủ thuật ký hiệu:** Viết $\mu^x(1-\mu)^{1-x}$ là cách gọn để biểu diễn: khi $x=1$ cho $\mu$, khi $x=0$ cho $(1-\mu)$. Thủ thuật này cũng dùng trong Multinomial.

### Phân Phối Binomial

Số lần $X=1$ xuất hiện trong $N$ phép thử Bernoulli:

$$p(m \mid N, \mu) = \binom{N}{m}\mu^m(1-\mu)^{N-m} \tag{6.95}$$

$$\mathbb{E}[m] = N\mu, \quad \mathbb{V}[m] = N\mu(1-\mu) \tag{6.96, 6.97}$$

Ứng dụng: xác suất quan sát $m$ lần "mặt ngửa" trong $N$ lần tung đồng xu.

### Phân Phối Beta

Phân phối liên tục trên $\mu \in [0,1]$, dùng để mô hình hóa **xác suất của một sự kiện nhị phân**:

$$p(\mu \mid \alpha, \beta) = \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}\mu^{\alpha-1}(1-\mu)^{\beta-1} \tag{6.98}$$

$$\mathbb{E}[\mu] = \frac{\alpha}{\alpha+\beta}, \quad \mathbb{V}[\mu] = \frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)} \tag{6.99}$$

trong đó $\Gamma(t) = \int_0^\infty x^{t-1}e^{-x}dx$, $\Gamma(t+1) = t\Gamma(t)$ — tổng quát hóa giai thừa.

**Các trường hợp đặc biệt:**

| $\alpha, \beta$ | Hình dạng phân phối |
|---|---|
| $\alpha = \beta = 1$ | Phân phối đều $U[0,1]$ |
| $\alpha, \beta < 1$ | Bimodal, spike ở 0 và 1 |
| $\alpha, \beta > 1$ | Unimodal |
| $\alpha = \beta > 1$ | Unimodal, đối xứng, tâm ở $\frac{1}{2}$ |

> **Trực quan:** $\alpha$ kéo xác suất về phía 1, $\beta$ kéo về phía 0.

---

## 6.6.1 Prior Liên Hợp (Conjugate Prior)

### Lý thuyết

**Định nghĩa 6.13:** Prior $p(\theta)$ là **conjugate prior** cho likelihood $p(x|\theta)$ nếu posterior $p(\theta|x)$ có **cùng dạng hàm** với prior.

> **Tại sao quan trọng?** Với conjugate prior, ta tính posterior bằng cách **cập nhật tham số** của prior — không cần tính tích phân phức tạp.

### Ví dụ 6.11 — Beta-Binomial Conjugacy

Prior: $\mu \sim \text{Beta}(\alpha, \beta)$, likelihood: $x \sim \text{Bin}(N, \mu)$.

Sau khi quan sát $h$ lần "heads" trong $N$ lần tung:

$$p(\mu \mid x=h) \propto p(x \mid \mu) p(\mu) \propto \mu^h(1-\mu)^{N-h} \cdot \mu^{\alpha-1}(1-\mu)^{\beta-1}$$

$$= \mu^{h+\alpha-1}(1-\mu)^{(N-h)+\beta-1} \propto \text{Beta}(h+\alpha,\ N-h+\beta) \tag{6.104}$$

Posterior vẫn là Beta — **chỉ cộng thêm dữ liệu quan sát vào tham số!**

> **Ý nghĩa:** $\alpha$ và $\beta$ đóng vai trò như số lần "heads" và "tails" đã quan sát trước đó (pseudo-counts). Quan sát thêm $h$ heads và $(N-h)$ tails → cập nhật tương ứng.

### Ví dụ 6.12 — Beta-Bernoulli Conjugacy

Prior: $\theta \sim \text{Beta}(\alpha, \beta)$, likelihood: $x \sim \text{Ber}(\theta)$.

$$p(\theta \mid x) \propto \theta^x(1-\theta)^{1-x} \cdot \theta^{\alpha-1}(1-\theta)^{\beta-1} = \theta^{\alpha+x-1}(1-\theta)^{\beta+(1-x)-1}$$

$$\propto \text{Beta}(\alpha+x,\ \beta+(1-x)) \tag{6.105}$$

Posterior vẫn là Beta. Nếu $x=1$ (quan sát heads): $\alpha \to \alpha+1$. Nếu $x=0$ (tails): $\beta \to \beta+1$.

### Bảng Prior Liên Hợp (Bảng 6.2)

| Likelihood | Conjugate Prior | Posterior |
|---|---|---|
| Bernoulli | Beta | Beta |
| Binomial | Beta | Beta |
| Gaussian (univariate) | Gaussian/Inverse-Gamma | Gaussian/Inverse-Gamma |
| Gaussian (multivariate) | Gaussian/Inverse-Wishart | Gaussian/Inverse-Wishart |
| Multinomial | Dirichlet | Dirichlet |

---

## 6.6.2 Thống Kê Đủ (Sufficient Statistics)

### Lý thuyết

**Thống kê đủ** $\phi(\mathbf{x})$ là tập thống kê chứa **toàn bộ thông tin** từ dữ liệu cần để suy luận về tham số $\theta$ — không mất thông tin nào khi thay $\mathbf{x}$ bằng $\phi(\mathbf{x})$.

**Định lý Fisher-Neyman:** $\phi(\mathbf{x})$ là thống kê đủ cho $\theta$ khi và chỉ khi:

$$p(\mathbf{x} \mid \theta) = h(\mathbf{x})\, g_\theta(\phi(\mathbf{x})) \tag{6.106}$$

trong đó $h(\mathbf{x})$ **không phụ thuộc** $\theta$, còn $g_\theta$ phụ thuộc $\theta$ **chỉ qua** $\phi(\mathbf{x})$.

> **Ví dụ:** Với $x_1, \ldots, x_N \sim \mathcal{N}(\mu, \sigma^2)$, thống kê đủ là $\{\bar{x}, \hat{\sigma}^2\}$ — toàn bộ thông tin về $(\mu, \sigma^2)$ được gói gọn trong hai số này, không cần giữ toàn bộ $N$ điểm dữ liệu.

---

## 6.6.3 Họ Hàm Mũ (Exponential Family)

### Lý thuyết

**Họ hàm mũ** là họ phân phối xác suất có dạng:

$$\boxed{p(\mathbf{x} \mid \boldsymbol{\theta}) = h(\mathbf{x})\exp\!\left(\langle\boldsymbol{\theta}, \phi(\mathbf{x})\rangle - A(\boldsymbol{\theta})\right)} \tag{6.107}$$

trong đó:
- $\phi(\mathbf{x})$: **vectơ thống kê đủ** (sufficient statistics)
- $\boldsymbol{\theta}$: **tham số tự nhiên** (natural parameters)
- $h(\mathbf{x})$: hàm cơ sở, không phụ thuộc $\boldsymbol{\theta}$
- $A(\boldsymbol{\theta})$: **log-partition function** — hằng số chuẩn hóa đảm bảo tích phân bằng 1

Dạng đơn giản (bỏ $h$ và $A$):

$$p(\mathbf{x} \mid \boldsymbol{\theta}) \propto \exp(\boldsymbol{\theta}^\top \phi(\mathbf{x})) \tag{6.108}$$

### Ví dụ 6.13 — Gaussian là Exponential Family

Với $\phi(x) = [x, x^2]^\top$:

$$p(x \mid \boldsymbol{\theta}) \propto \exp(\theta_1 x + \theta_2 x^2)$$

Đặt $\boldsymbol{\theta} = \left[\frac{\mu}{\sigma^2},\ -\frac{1}{2\sigma^2}\right]^\top$:

$$p(x \mid \boldsymbol{\theta}) \propto \exp\!\left(\frac{\mu x}{\sigma^2} - \frac{x^2}{2\sigma^2}\right) \propto \exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right) \tag{6.111}$$

**Kết quả:** Gaussian là exponential family với $\phi(x) = [x, x^2]^\top$ — **mean và variance là thống kê đủ**.

### Ví dụ 6.14 — Bernoulli là Exponential Family

$$p(x \mid \mu) = \mu^x(1-\mu)^{1-x} = \exp\!\left[x\log\frac{\mu}{1-\mu} + \log(1-\mu)\right] \tag{6.113}$$

Nhận dạng:
- $h(x) = 1$
- $\theta = \log\frac{\mu}{1-\mu}$ (log-odds)
- $\phi(x) = x$
- $A(\theta) = \log(1 + e^\theta)$

Mối quan hệ giữa $\theta$ và $\mu$:

$$\mu = \frac{1}{1+e^{-\theta}} \tag{6.118}$$

> **Kết nối quan trọng:** Đây chính là **hàm sigmoid** (logistic function)! Sigmoid xuất hiện tự nhiên như hàm liên kết giữa tham số tự nhiên và tham số gốc của Bernoulli — đây là lý do sigmoid được dùng trong logistic regression và làm activation function trong neural networks.

### Prior Liên Hợp của Exponential Family

Mọi phân phối trong exponential family đều có conjugate prior dạng:

$$p(\boldsymbol{\theta} \mid \boldsymbol{\gamma}) = h_c(\boldsymbol{\theta})\exp\!\left(\left\langle\begin{bmatrix}\gamma_1\\\gamma_2\end{bmatrix}, \begin{bmatrix}\boldsymbol{\theta}\\-A(\boldsymbol{\theta})\end{bmatrix}\right\rangle - A_c(\boldsymbol{\gamma})\right) \tag{6.120}$$

**Ví dụ 6.15:** Áp dụng công thức trên cho Bernoulli → thu được Beta distribution — **không cần đoán** dạng conjugate prior, mà **suy ra** từ dạng tổng quát!

---

## Tổng Kết Mục 6.6

- **Ba phân phối nền tảng:** Bernoulli (nhị phân), Binomial (đếm), Beta (xác suất liên tục) — được thiết kế để làm việc với nhau qua conjugacy.
- **Conjugate prior** = prior và posterior cùng dạng → cập nhật Bayesian chỉ là **cộng pseudo-counts vào tham số**.
- **Thống kê đủ** = tập con thông tin từ dữ liệu **chứa đủ** mọi thứ cần cho suy luận về $\theta$.
- **Exponential family** thống nhất hầu hết các phân phối có tên dưới một dạng chung, đảm bảo cả ba tiêu chí: tính đóng (conjugacy), số tham số hữu hạn (sufficient statistics), và tối ưu hóa thuận lợi (log-likelihood lõm).
- **Sigmoid function** xuất hiện tự nhiên từ exponential family của Bernoulli — kết nối xác suất học máy với logistic regression và neural networks.

# 6.7 Đổi Biến / Biến Đổi Nghịch (Change of Variables / Inverse Transform)

---

## Bối Cảnh

Các phân phối "có tên" rất hạn chế. Trong ML, ta thường cần **phân phối của biến đã biến đổi** — ví dụ: nếu $X \sim \mathcal{N}(0,1)$, phân phối của $X^2$ là gì? Có hai cách tiếp cận chính.

**Biến rời rạc** — đơn giản: Nếu $Y = U(X)$ và $U$ khả nghịch thì:

$$P(Y = y) = P(X = U^{-1}(y)) \tag{6.125}$$

**Biến liên tục** — phức tạp hơn vì $P(Y=y) = 0$ tại mọi điểm, phải làm việc với pdf. Ta có hai kỹ thuật.

---

## 6.7.1 Kỹ Thuật Hàm Phân Phối (Distribution Function Technique)

### Lý thuyết

Tìm pdf của $Y = U(X)$ qua hai bước:

**Bước 1:** Tính cdf của $Y$:
$$F_Y(y) = P(Y \leq y) \tag{6.126}$$

**Bước 2:** Lấy đạo hàm cdf để ra pdf:
$$f(y) = \frac{d}{dy}F_Y(y) \tag{6.127}$$

### Ví dụ 6.16 — Tìm pdf của $Y = X^2$

Cho $X$ có pdf $f(x) = 3x^2$ trên $[0,1]$. Tìm pdf của $Y = X^2$.

**Bước 1:** Tính cdf của $Y$:

$$F_Y(y) = P(Y \leq y) = P(X^2 \leq y) = P(X \leq y^{1/2}) = F_X(y^{1/2}) = \int_0^{y^{1/2}} 3t^2\,dt = y^{3/2} \tag{6.129}$$

**Bước 2:** Lấy đạo hàm:

$$f(y) = \frac{d}{dy}F_Y(y) = \frac{3}{2}y^{1/2}, \quad 0 \leq y \leq 1 \tag{6.131}$$

---

### Định Lý 6.15 — Probability Integral Transform

Nếu $X$ là biến ngẫu nhiên liên tục với cdf $F_X(x)$ đơn điệu chặt, thì:

$$Y := F_X(X) \sim U[0,1] \tag{6.132}$$

> **Ứng dụng lấy mẫu:** Đảo ngược định lý này — để lấy mẫu từ phân phối bất kỳ có cdf $F_X$:
> 1. Lấy mẫu $u \sim U[0,1]$
> 2. Tính $x = F_X^{-1}(u)$
>
> Đây là **inverse CDF sampling** — nền tảng của nhiều thuật toán lấy mẫu.

---

## 6.7.2 Kỹ Thuật Đổi Biến (Change of Variables)

### Lý thuyết — Một chiều

Cho $X$ với pdf $f_X(x)$ và hàm khả nghịch $U$, đặt $Y = U(X)$. Từ định nghĩa cdf và chain rule:

$$\boxed{f(y) = f_X(U^{-1}(y)) \cdot \left|\frac{d}{dy}U^{-1}(y)\right|} \tag{6.143}$$

**Giải thích:** Hệ số $\left|\frac{d}{dy}U^{-1}(y)\right|$ đo **hệ số co giãn thể tích** khi áp dụng $U$ — tương tự định thức Jacobian trong Mục 5.3.

**Tại sao cần trị tuyệt đối?** Nếu $U$ giảm thì đạo hàm âm, nhưng pdf phải dương. Trị tuyệt đối đảm bảo điều này cho cả hàm tăng và giảm.

> **So sánh với rời rạc:** Trường hợp rời rạc (6.125b) không có hệ số $\left|\frac{d}{dy}U^{-1}(y)\right|$ vì xác suất tại điểm có nghĩa. Với liên tục, pdf không phải xác suất tại điểm nên cần điều chỉnh thêm co giãn.

---

### Định Lý 6.16 — Đổi Biến Đa Chiều

Cho $f_X(\mathbf{x})$ là pdf của $X \in \mathbb{R}^D$ và $\mathbf{y} = U(\mathbf{x})$ khả vi và khả nghịch. Khi đó pdf của $Y = U(X)$:

$$\boxed{f(\mathbf{y}) = f_X(U^{-1}(\mathbf{y})) \cdot \left|\det\!\left(\frac{\partial}{\partial \mathbf{y}}U^{-1}(\mathbf{y})\right)\right|} \tag{6.144}$$

**Thay trị tuyệt đối đạo hàm bằng trị tuyệt đối định thức Jacobian** — vì trong nhiều chiều, "hệ số co giãn thể tích" của một ánh xạ chính là $|\det(J)|$ (nhớ lại Mục 5.3 và Mục 4.1).

**Quy trình áp dụng:**
1. Tìm biến đổi ngược $U^{-1}(\mathbf{y})$
2. Thay vào $f_X$ để được $f_X(U^{-1}(\mathbf{y}))$
3. Tính Jacobian $\frac{\partial}{\partial \mathbf{y}}U^{-1}(\mathbf{y})$ và lấy $|\det|$
4. Nhân hai kết quả lại

---

### Ví dụ 6.17 — Biến Đổi Tuyến Tính của Gaussian Chuẩn Tắc

Cho $X \sim \mathcal{N}(\mathbf{0}, I)$ với pdf:

$$f\!\left(\begin{bmatrix}x_1\\x_2\end{bmatrix}\right) = \frac{1}{2\pi}\exp\!\left(-\frac{1}{2}\begin{bmatrix}x_1\\x_2\end{bmatrix}^\top\begin{bmatrix}x_1\\x_2\end{bmatrix}\right) \tag{6.145}$$

Cho $A = \begin{bmatrix}a&b\\c&d\end{bmatrix}$, tìm pdf của $Y = AX$.

**Bước 1 — Biến đổi ngược:**

$$\mathbf{x} = A^{-1}\mathbf{y} = \frac{1}{ad-bc}\begin{bmatrix}d&-b\\-c&a\end{bmatrix}\mathbf{y} \tag{6.147}$$

**Bước 2 — Thay vào pdf của $X$:**

$$f_X(A^{-1}\mathbf{y}) = \frac{1}{2\pi}\exp\!\left(-\frac{1}{2}\mathbf{y}^\top A^{-\top}A^{-1}\mathbf{y}\right) \tag{6.148}$$

**Bước 3 — Tính Jacobian** (từ Mục 5.5: $\frac{\partial}{\partial \mathbf{y}}A^{-1}\mathbf{y} = A^{-1}$):

$$\det\!\left(\frac{\partial}{\partial \mathbf{y}}A^{-1}\mathbf{y}\right) = \det(A^{-1}) = \frac{1}{ad-bc} \tag{6.150}$$

**Bước 4 — Áp dụng công thức (6.144):**

$$f(\mathbf{y}) = \frac{1}{2\pi}\exp\!\left(-\frac{1}{2}\mathbf{y}^\top A^{-\top}A^{-1}\mathbf{y}\right) \cdot |ad-bc|^{-1} \tag{6.151}$$

> **Kết nối với Gaussian:** $A^{-\top}A^{-1} = (AA^\top)^{-1} = \Sigma^{-1}$ nên $f(\mathbf{y})$ chính là $\mathcal{N}(\mathbf{y} \mid \mathbf{0}, AA^\top) = \mathcal{N}(\mathbf{y} \mid \mathbf{0}, \Sigma)$ — nhất quán với kết quả biến đổi tuyến tính Gaussian ở Mục 6.5.3.

---

## Tổng Kết Mục 6.7

| Kỹ thuật | Cách làm | Khi nào dùng |
|---|---|---|
| **Distribution function** | cdf → đạo hàm → pdf | Trực tiếp từ định nghĩa, rõ ràng |
| **Change of variables** | Jacobian + inverse transform | Tổng quát hơn, dùng được đa chiều |

- **Một chiều:** $f(y) = f_X(U^{-1}(y)) \cdot \left|\frac{d}{dy}U^{-1}(y)\right|$ — hệ số điều chỉnh là đạo hàm của hàm ngược
- **Đa chiều:** thay $\left|\cdot\right|$ bằng $|\det(J)|$ — định thức Jacobian đo hệ số co giãn thể tích
- **Probability integral transform:** $F_X(X) \sim U[0,1]$ → nền tảng của inverse CDF sampling
- Kỹ thuật này xuất hiện trong **reparametrization trick** (VAE), **normalizing flows**, và đổi biến trong tích phân xác suất
