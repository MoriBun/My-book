- Giới thiệu chung: đại số là xây dựng một tập hợp các đối tượng và các quy tắc để thao tác. Đại số tuyến tính nghiên cứu vector và các quy tắc cộng/trái vô hướng trên vector.

- Định nghĩa vector: vector là các đối tượng có thể cộng với nhau và nhân với số vô hướng để vẫn thu được đối tượng cùng loại.

- Ví dụ về vector:
    1. Vector hình học: các đoạn có hướng, có thể cộng và nhân với số vô hướng. Đây là hình ảnh trực quan quen thuộc.
    2. Đa thức: hai đa thức có thể cộng với nhau và nhân với số vô hướng, nên cũng được xem là vector dù trừu tượng hơn.
    3. Tín hiệu âm thanh: dãy số, có thể cộng và nhân vô hướng, nên là vector.
    4. Phần tử của R^n: bộ n số thực, là dạng vector được tập trung trong sách này.

- Tập trung vào R^n vì nó phù hợp với thuật toán và gần với mảng số trên máy tính; nhiều ngôn ngữ lập trình hỗ trợ phép toán mảng.

- Ý tưởng chính: đóng hợp (closure) — tập các đối tượng thu được khi cộng và nhân vô hướng từ một tập ban đầu. Tập đó tạo thành không gian vector.

- Mục đích: nắm các khái niệm cơ bản về vector, không gian vector, và mối liên hệ với các chương sau như hình học, đạo hàm vectơ, giảm chiều, hồi quy tuyến tính.
